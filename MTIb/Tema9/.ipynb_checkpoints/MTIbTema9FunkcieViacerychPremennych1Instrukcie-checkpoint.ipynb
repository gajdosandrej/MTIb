<a id=motivacia></a>
# <font color=brown> Motivácia</font>


 

Doteraz sme sa zaoberali predovšetkým reálnymi funkciami jednej reálnej premennej a skúmali sme ich pomocou rôznych nástrojov diferenciálneho a integrálneho počtu. V skutočnom svete však existujú veličiny a vzťahy medzi nimi, ktoré nie je možné popísať len takými funkciami ale potrebujeme k tomu viacero premenných. Napr. **keď sa webová stránka pokúša "predpovedať" správanie používateľa**, môže vytvoriť funkciu, ktorá bude obsahovať tisíce premenných vrátane veku používateľa, súradníc jeho polohy, počtu kliknutí na odkazy určitého typu atď. lebo to, čo uživateľ urobí (výstup) závisí od **viacerých premenných** (veličín). Výstup môže obsahovať jednu ale aj viac premenných (hodnôt), napríklad pravdepodobnosť kliknutia na iný odkaz alebo pravdepodobnosť zakúpenia inej položky a pod. 

 

Funkcie tohto typu, kde na vstupe je niekoľko reálnych premenných a výstupom je jedna reálna hodnota, budú náplňou nasedujúcich tém v rámci MTIb. Ide o tzv. **funkcie viacerých premenných**. V téme 9 sa pozrieme na konkrétne **príklady** takých **funkcií**, ukážeme si **spôsoby** ich **vizualizácie** a tiež sa budeme zaoberať určovaním ich **definičných oborov**. Budeme sa venovať najmä **funkciám dvoch** alebo **troch premenných**, keďže tieto ešte vieme aj celkom rozumne znázorniť, no niektoré definície či vlastnosti budú formulované i všeobecnejšie, tzn. pre funkcie **$n$ premenných**.

<a id=ciele></a>
# <font color=brown> Učebné ciele</font>

### Základné 


* Uviesť konkrétne príklady funkcií viacerých premenných. 


* Definovať funkciu $n$ premenných, špeciálne pre $n=2$, $n=3$ definovať graf a vrstevnicový diagram (úrovňové plochy), ilustrovať na príkladoch. 


* Určiť vzdialenosť dvoch bodov v 2D aj v 3D. 


* Rozumieť pojmu funkcia daná implicitne a uviesť príklad. 


* Definovať lineárnu funkciu, ilustrovať na príklade a vedieť zobraziť. 

### Pokročilé 


* Znázorniť grafy aj vrstevnicové diagramy funkcie dvoch premenných pomocou Sage. 


* S využitím Sage určiť a znázorniť definičné obory funkcie dvoch premenných. 


* Vizualizovať plochy (napr. dané implicitnou funkciou) v 3D pomocou Sage. 


* Nájsť predpis lineárnej funkcie s danými vlastnosťami (napr. podľa tabuľky s kontextom alebo podľa vrstevníc) a vedieť ju vizualizovať. 


* Určiť, či graf danej plochy v 3D resp. krivky v 2D môže byť grafom funkcie dvoch resp. jednej premennej a nájsť predpis takej funkcie, pokiaľ je to možné. 


* Vedieť identifikovať základné kvadratické plochy podľa rovnice s využitím galérie kvadratických plôch.
