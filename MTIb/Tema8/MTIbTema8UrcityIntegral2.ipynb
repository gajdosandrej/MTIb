# <font color=brown> Výpočet obsahov rovinných útvarov </font>

Uvažujme najprv **útvar ohraničený grafom spojitej funkcie $f$, osou $x$ a priamkami $x=a,x=b$, pričom $f$ je na danom intervale kladná**. Potom vzhľadom na interpretáciu určitého integrálu ako plošného obsahu dostávame, že uvedený útvar má obsah rovný $\displaystyle\int\limits_a^b f(x)\,\mathrm{d}x$.
 
Ak **spojitá funkcia $f$ nadobúda na intervale $\langle a,b\rangle$ záporné hodnoty**, tak (vzhľadom na symetriu podľa osi $x$) plošný obsah zodpovedajúceho útvaru je rovný plošnému obsahu útvaru, ktorý je ohraničený grafom funkcie $-f$, osou $x$ a priamkami $x=a,x=b$, čiže je rovný $-\displaystyle\int\limits_a^b f(x)\,\mathrm{d}x$. 


$$\text{Zdroj: Madaras (2020).}$$ 

Ak **funkcia $f$ je spojitá na $\langle a,b\rangle$, ale na niektorých častiach intervalu $\langle a,b\rangle$ je kladná a na iných je záporná**, tak obsah $P$ obrazca ohraničeného priamkami $x=a$, $x=b$, $y=0$ a krivkou $y=f(x)$ vypočítame ako súčet plôch. Napríklad pri funkcii znázornenej na obrázku


$$\text{Zdroj: Madaras (2020).}$$

je obsah obrazca $P=P_1+P_2+P_3=\displaystyle\int\limits^c_af(x)\,\mathrm{d}x-\displaystyle\int\limits^d_cf(x)\,\mathrm{d}x+\displaystyle\int\limits^b_df(x)\,\mathrm{d}x$. 

Ak je **útvar ohraničený priamkami $x=a$, $x=b$ a krivkami $y=f(x)$, $y=g(x)$, pričom $f$, $g$ sú spojité na $\langle a,b\rangle$ a platí $f(x)\ge g(x)$ pre všetky $x\in\langle a,b\rangle$** (bez ohľadu na to, či hodnoty funkcií $f,g$ sú kladné alebo záporné), tak jeho obsah je rovný $\displaystyle\int\limits_a^b \left(f(x) - g(x)\right)\,\mathrm{d}x$.


$$\text{Zdroj: Madaras (2020).}$$

Ak je **útvar ohraničený krivkami spojitých funkcií $f,g$ a jedinou priamkou $x = a$**, tak je najprv potrebné určiť spoločný bod kriviek $y = f(x),y = g(x)$ (t.j. bod $x^*$, pre ktorý $f(x^*) = g(x^*)$). Ďalej potrebujeme stanoviť, či na intervale $\langle x^*,a \rangle$ (resp. $\langle a,x^*\rangle$) platí $f(x) \leq g(x)$ alebo $g(x) \leq f(x)$ (predpokladáme, že na daných intervaloch je jeden z grafov $f,g$ vždy "nad" tým druhým). V prípade, že $x^* < a$ je plošný obsah daného útvaru rovný $\displaystyle\int\limits_{x^*}^a (f(x) - g(x))\, \mathrm{d}x$ (ak $f(x) \geq g(x)$ na $\langle x^*,a \rangle$) resp. $\displaystyle\int\limits_{x^*}^a (g(x) - f(x))\, \mathrm{d}x$ (ak $g(x) \geq f(x)$ na $\langle x^*,a \rangle$); ak $x^* > a$, tak v uvedených určitých integráloch vymeníme integračné hranice.

Podobne postupujeme aj v prípade, keď je **útvar ohraničený len krivkami $y = f(x),y = g(x)$**: nájdeme dva rôzne body $x^*,x^{+}$ také, že $x^* < x^{+}$ a $f(x^*) = g(x^*),f(x^+) = g(x^+)$, teda body, kde sa pretínajú grafy funkcií $f,g$; plošný obsah útvaru je potom $\displaystyle\int\limits_{x^*}^{x^+} (f(x) - g(x))\, \mathrm{d}x$ resp. $\displaystyle\int\limits_{x^*}^{x^+} (g(x) - f(x))\, \mathrm{d}x$ podľa toho, či na intervale $\langle x^*,x^+ \rangle$ je graf funkcie $g$ pod, alebo nad grafom funkcie $f$.

#### <font color=blue> Príklad 1 </font>  

Vypočítajte obsah obrazca ohraničeného krivkami $y=x^2$, $y^2=x$.  

<ins> Riešenie: </ins> 

Všimnime si, že vzťah $y^2=x$ nie je vzťah vyjadrujúci funkciu: máme totiž $y=\pm\sqrt x$. Obrazec vyzerá nasledovne:


$$\text{Zdroj: Madaras (2020).}$$

Z náčrtu možno povedať, že daný útvar je na $\langle0,1\rangle$ ohraničený funkciami $f: y=\sqrt x$ a $g: y=x^2$; súčasne platí $f(x)\ge g(x)$ pre všetky $x\in\langle0,1\rangle$. Plošný obsah útvaru potom vypočítame ako určitý integrál rozdielu funkcií $f$ a $g$ v hraniciach daných spoločnými bodmi ich grafov:

$$P=\displaystyle\int\limits^1_0\left(\sqrt x-x^2\right)\,\mathrm{d}x=\left[\frac{x^{\frac32}}{\frac32}-\frac{x^3}3\right]^1_0=\frac23-\frac13=\frac13\,.$$


```python
# obrazok v Sage, nas ale zaujima iba interval od 0 po 1  
x = var('x') 
f(x) = sqrt(x) 
g(x) = x^2 
plot([f,g], (x,0,2), color=['red','green'], fill=g, fillcolor='gray') + text('$y=x^2$', (1.5, 3), color='green', fontsize=14, figsize=5) + text('$y^2=x$', (1.5, 0.8), color='red', fontsize=14)
```

<ins> Riešenie: </ins>

$$\text{Zdroj: Madaras (2020).}$$ 

Plocha prvého útvaru (ohraničeného krivkami $y=x^3+1, y = x$ a priamkami $x = 0,x=2$) je

# <font color=brown> Výpočet objemov rotačných telies </font>


Ak rotačné teleso vznikne tak, že útvar ohraničený priamkami $x=a$, $x=b$, osou $x$ a krivkou $y=f(x)$ rotuje okolo osi $x$, tak jeho objem je rovný $\pi\displaystyle\int\limits_a^b f^2(x)\, \mathrm{d}x$. 


$$\text{Zdroj: Madaras (2020).}$$ 

Ak rotačné teleso vznikne rotáciou (okolo osi $x$) útvaru  ohraničeného priamkami $x=a$, $x=b$ a dvojicou kriviek $y = f(x), y = g(x)$, pričom funkcie $f,g$ sú spojité a pre každé $x \in \langle a,b \rangle$ platí $f(x) \geq g(x) \geq 0$, tak jeho objem je rovný $\pi\displaystyle\int\limits_a^b (f^2(x) - g^2(x))\,\mathrm{d}x$. 


$$\text{Zdroj: Madaras (2020).}$$

#### <font color=blue> Príklad 3 </font>  

Vypočítajte objem telesa, ktoré vznikne otáčaním okolo osi $x$ obrazca ohraničeného krivkami $y=x^2+1$, $y=x+3$.   

<ins> Riešenie: </ins> 

Spoločné body kriviek $y =x^2+1, y=x+3$ spĺňajú rovnicu $x^2 + 1 = x+3$, ktorej riešením je $x_1=-1$, $x_2=2$. Potom objem uvedeného rotačného telesa je 

$$
\begin{aligned}
	V&=\pi\int\limits^2_{-1}\bigg((x+3)^2-\big(x^2+1\big)^2\bigg)\,\mathrm{d}x\\
	&=\pi\int\limits^2_{-1}\big(x^2+6x+9-x^4-2x^2-1\big)\,\mathrm{d}x=\\
	&=\pi\int\limits^2_{-1}\big(-x^2+6x+8-x^4\big)\,\mathrm{d}x=\pi\bigg[-\frac{x^3}3+3x^2+8x-\frac{x^5}5\bigg]^2_{-1}=\\
	&=\pi\bigg[-\frac83+12+16-\frac{32}{5}-\bigg(\frac13+3-8+\frac15\bigg)\bigg]=\frac{117}{5}\pi.
\end{aligned}
$$


```python
# nas zaujima rotacia (okolo osi x) plochy medzi x=-1 a x=2
x = var('x') 
f(x) = x^2+1 
g(x) = x+3 
plot([f,g], (x,-2,3), color=['red','green'], fill=g, fillcolor='gray', figsize=5)
```

# <font color=brown> Výpočet dĺžok kriviek </font>



Uvažujme graf funkcie $y=\sin x$ na intervale $x\in\langle 0,\pi \rangle$ a položme si otázku, aká dlhá je krivka odpovedajúca grafu tejto funkcie na zvolenom intervale? Inými slovami, ak by sme použili špagát, ktorý by presne zodpovedal tvaru tejto krivky (prekryl by ju), aký dlhý by bol tento špagát? 

Ako sme to už urobili v minulosti, **začneme aproximáciou a neskôr svoju odpoveď spresníme pomocou limít, aby sme dostali presné riešenie nastoleného problému**. Dĺžka úsečky sa dá ľahko vypočítať pomocou vzorca pre vzdialenosť dvoch bodov. Dĺžku danej krivky môžeme približne určiť tak, že krivku aproximujeme niekoľkými úsečkami t.j. rozdelíme ju na konečný počet úsečiek (na intervale $\langle 0,\pi \rangle$) a sčítaním dĺžok týchto úsečiek dostaneme odhad dĺžky pôvodnej krivky, ktorá predstavuje graf funkcie $y=\sin x$ na intervale $x\in\langle 0,\pi \rangle$.

Na nasledujúcom obrázku je krivka aproximovaná štyrmi úsečkami (znázornené červenou farbou) - interval $x\in\langle 0,\pi \rangle$ je rovnomerne rozdelený na štyri podintervaly. 


$$\text{Zdroj: Hartman G. et al. (2018).}$$

Z obrázka je vidieť, že tieto úsečky aproximujú krivku (graf funkcie $y=\sin x$) na prvom aj štvrtom intervale vcelku presne, avšak na druhom a treťom intervale je aproximácia o čosi horšia (úsečky nepriliehajú dostatočne ku grafu funkcie). Ak by sme spočítali dĺžky všetkých štyroch úsečiek, dostali by sme hodnotu $3.79$, čo je približná hodnota dĺžky krivky. Zrejme, ak by sme miesto štyroch úsečiek aproximovali krivku napr. šiestimi úsečkami, tak táto aproximácia by bola presnejšia a teda čím vyšší počet úsečiek, tým lepšia aproximácia krivky a tým presnejší odhad dĺžky krivky. 

Vo všeobecnosti teda môžeme aproximovať dĺžku krivky $y=f(x)$ na intervale $\langle a,b \rangle$ nasledovne: Nech $a=x_0<x_1<\ldots<x_{n-1}<x_{n}=b$ je delenie intervalu $\langle a,b \rangle$ na $n$ podintervalov. Nech $\Delta x_i$ reprezentuje dĺžku $i$-teho intervalu (pre jednoduchosť by sme mohli uvažovať rovnomerné delenie, kde $\Delta x_i=\Delta x$). Na $i$-tom podintervale je krivka aproximovaná úsečkou (znázornená červenou farbou) tak, ako je to znázornené na nasledujúcom obrázku. Dĺžku takej úsečky (znázornená červenou farbou) môžeme určiť ako dĺžku prepony v pravouhlom trojuholníku (viď obrázok). 


$$\text{Zdroj: Hartman G. et al. (2018).}$$

Pre jednoduchosť uvažujme rovnomerné delenie intervalu $\langle a,b \rangle$, podobne ako pri zavádzaní určitého integrálu. 
Ak máme peknú "hladkú" krivku (ako na obrázku), tak malej zmene $\Delta x$ zodpovedá malá zmena $\Delta y_i$ a z diferenciálneho počtu (Lagrangeova veta o strednej hodnote) vieme, že $\Delta y_i$ môžeme nahradiť výrazom $f'(t_i)\Delta x$, kde $t_i$ je bod z intervalu $(x_{i-1},x_i)$. Dĺžku $i$-tej úsečky potom môžeme prepísať do tvaru $\sqrt{\Delta x_i^2+\Delta y_i^2}=\sqrt{1+f'(t_i)}\Delta x$. **Sčítaním dĺžok všetkých $n$ úsečiek dostaneme $\displaystyle\sum\limits_{i=1}^{n}\sqrt{1+\left[f'(t_i)\right]^2}\Delta x$, čo je približná dĺžka pôvodnej krivky danej funkciou $y=f(x)$ na intervale $\langle a,b \rangle$.** 

Zároveň to, čo sme dostali, je **Riemannov súčet funkcie $\sqrt{1+\big[f'(x)\big]^2}$**. Ak teda **urobíme z tohto Riemannovo súčtu $\displaystyle\sum\limits_{i=1}^{n}\sqrt{1+\left[f'(t_i)\right]^2}\Delta x$ limitu pre $n\to\infty$**, tzn. budeme zjemňovať delenie (pridávať stále viac a viac deliacich bodov resp. úsečiek, tým pádom $\Delta x\to0$), **potom dostaneme skutočnú dĺžku krivky a jej hodnota bude rovná $\displaystyle\int\limits^b_a\sqrt{1+\big[f'(x)\big]^2}\,\mathrm{d}x$ (suma sa "zmení" na integrál pri $n\to\infty$)**.


**Ak je krivka (t.j. časť grafu funkcie) daná rovnicou $y = f(x)$ pre $x \in \langle a,b \rangle$, tak dĺžka krivky medzi bodmi $[a,f(a)]$ a $[b,f(b)]$ je rovná $\displaystyle\int\limits^b_a\sqrt{1+\big[f'(x)\big]^2}\,\mathrm{d}x$.**

#### <font color=blue> Príklad 4 </font>  

Vypočítajte dĺžku krivky $f: y = x^{\frac{3}{2}}$ medzi bodmi s $x$-ovými súradnicami 0 a 1.   

<ins> Riešenie: </ins> 

Platí $f'(x) = \frac{3}{2}x^{\frac{1}{2}} = \frac{3}{2}\sqrt{x}$; dĺžka danej krivky bude teda rovná

$$l = \displaystyle\int\limits_0^1 \sqrt{1+\big[f'(x)\big]^2}\,\mathrm{d}x = \displaystyle\int\limits_0^1 \sqrt{1 + \left ( \frac{3}{2}\sqrt{x} \right )^2}\, \mathrm{d}x =\displaystyle\int\limits_0^1 \sqrt{1 + \frac{9}{4}x}\, \mathrm{d}x = \displaystyle\int\limits_0^1 \frac{\sqrt{4+9x}}{2}\,\mathrm{d}x = \frac{1}{2}\displaystyle\int\limits_0^1 \sqrt{4+9x}\, \mathrm{d}x.$$

Neurčitý integrál $\displaystyle\int\sqrt{4+9x}\,\mathrm{d}x$ vypočítame pomocou substitúcie $4+9x = t$, z ktorej máme $9\,\mathrm{d}x = \mathrm{d}t$, teda $\mathrm{d}x = \frac{\mathrm{d}t}{9}$ a

$$\displaystyle\int\sqrt{4+9x}\,\mathrm{d}x = \displaystyle\int \sqrt{t}\frac{\mathrm{d}t}{9} = \frac{1}{9} \displaystyle\int t^{\frac{1}{2}}\, \mathrm{d}t = \frac{1}{9}\cdot\frac{t^{\frac{3}{2}}}{\frac{3}{2}} = \frac{2}{27}\sqrt{t^3} = \frac{2}{27}\sqrt{(4+9x)^3}$$

Dostávame teda

$$l = \frac{1}{2}\displaystyle\int\limits_0^1 \sqrt{4+9x}\, \mathrm{d}x = \frac{1}{2}\left [ \frac{2}{27}\sqrt{(4+9x)^3} \right ]_0^1 = \frac{1}{27} (\sqrt{13^3} - \sqrt{4^3}) = \frac{13\sqrt{13} - 8}{27} \doteq 1.4397.$$


```python
x = var('x') 
f(x) = x^(3/2) 
plot(f, (x,0,1), figsize=4)
```

# <font color=brown> Nevlastné integrály </font>

Pri zavádzaní určitého integrálu sme predpokladali, že máme uzavretý interval (konečnej dĺžky) $\langle a,b \rangle$ a na ňom spojitú funkciu (o ktorej vieme vďaka Weierstrassovej vete, že je na tomto intervale ohraničená). Integrály, ktoré vznikajú v aplikáciách, nemusia mať tieto pekné vlastnosti. V tejto časti budeme skúmať triedu integrálov nazývanú **_nevlastné integrály_**, kde buď jedna z integračných hraníc (horná alebo dolná) je nekonečno alebo integrand je neohraničená funkcia. Napr. pri odhade hmotnosti zemskej atmosféry môžeme vypočítať integrál, ktorý kumuluje (nasčíta) hmotnosť vzduchu v rôznych výškach. Aby sme podchytili skutočnosť, že atmosféra nekončí v konkrétnej výške, hornú hranicu integrácie zväčšujeme stále viac  resp. ju môžeme nechať "ísť" do nekonečna. Uvažovať budeme predovšetkým nevlastné integrály s kladným integrandom, pretože sú najbežnejšie. 

### Nevlastné integrály s nekonečnou hranicou integrácie 

--- 

#### <font color=blue> Príklad 5 </font> 

Uvažujme nasledovný príklad nevlastného integrálu s nekonečnou hornou hranicou integrácie 

$$\displaystyle\int_{1}^{\infty}\frac{1}{x^2}\,\mathrm{d}x.$$ 

Ak nevieme hneď vyriešiť tento nevlastný integrál, tak to skúsme previesť na niečo, čo riešiť vieme. Pozrime sa, ako by vyzerali hodnoty určitých integrálov s konečnou, no stále sa zväčšujúcou hornou hranicou 

$$\displaystyle\int_{1}^{10}\frac{1}{x^2}\,\mathrm{d}x = \left[-\frac{1}{x}\right]^{10}_{1}=-\frac{1}{10}+1=0.9;$$ 

$$\displaystyle\int_{1}^{100}\frac{1}{x^2}\,\mathrm{d}x = 0.99;$$ 

$$\displaystyle\int_{1}^{1000}\frac{1}{x^2}\,\mathrm{d}x = 0.999.$$ 

Nasledujúci Sage kód dáva hodnoty pre $\displaystyle\int_{1}^{b}\frac{1}{x^2}\,\mathrm{d}x$, pričom $b\in\{10,100,1000,10000,100000,1000000\}$. 

```python
x = var('x')
f(x) = 1/x^2
[integral(f(x),x,1,b) for b in [10,100,1000,10000,100000,1000000]]
```

```python
x = var('x')
f(x) = 1/x^2
P = line2d([(1,0),(1,f(1))], color='black') 
P += line2d([(10,0),(10,f(10))], color='black')
P += polygon([(1,0),(1,f(1))] + [(x, f(x)) for x in [1.05,1.10,..,10]] + [(10,0),(1,0)],  rgbcolor=(0.8,0.8,0.8),aspect_ratio='automatic')
P += text("$\\int_{1}^{10} \\frac{1}{x^2} \\, \\mathrm{d}x \\, = \\, 0.9$", (2.5, 1), fontsize=12, color='black')
P += text("$f(x)=\\dfrac{1}{x^2}$", (1.8, 1.8), fontsize=12, color='blue')
P += plot(f, (0.5, 10), thickness=3) 
P
```

Z výpočtov vidíme, že nevlastný integrál by mohol nadobúdať určitú konečnú hodnotu lebo pri posúvaní hornej integračnej hranice smerom do nekonečna sa hodnoty jednotlivých integrálov blížia k hodnote $1$.

Keďže hodnota nevlastného integrálu $1$ je reálne číslo (konečné), tak hovoríme, že nevlastný integrál $\displaystyle\int_{1}^{\infty}\frac{1}{x^2}\,\mathrm{d}x$ _konverguje_ k $1$.

### <font color=red> Definícia 2 </font> 

Nech $f(x)$ je kladná a spojitá funkcia pre $a\le x < b$ a nech $f(x)\to\infty$ pre $x\to b$. Ak $\lim\limits_{c\to b^-}\displaystyle\int_{a}^{c}f(x)\,\mathrm{d}x$ je (konečné) reálne číslo, potom hovoríme, že $\displaystyle\int_{a}^{b}f(x)\,\mathrm{d}x$ **_konverguje_** a definujeme

Ináč hovoríme, že $\displaystyle\int_{a}^{b}f(x)\,\mathrm{d}x$ **_diverguje_**.

#### Poznámka 2

Keď $f(x)\to\infty$ pre $x\to a$, definujeme konvergenciu nevlastného integrálu analogicky ako v Definícii 2.

V skutočnosti, **integrál môže byť nevlastný aj kvôli tomu, že integrand sa blíži do nekonečna v blízkosti bodu, ktorý je vo vnútri zvoleného intervalu**, tzn. nie je to krajný bod intervalu. V takom prípade je možné rozdeliť daný integrál na dva (alebo viac) nevlastné integrály tak, aby integrand smeroval do nekonečna iba v nejakom krajnom bode intervalu, cez ktorý sa integruje resp. v blízkosti niektorej integračnej hranice.  

Predpokladajme, že $f(x)$ je kladná a spojitá na intervale $\langle a,b \rangle$ s výnimkou bodu $c\in(a,b)$. Ak $f(x)\to\infty$ pre $x\to c$, potom definujeme 

$$\displaystyle\int_{a}^{b}f(x)\,\mathrm{d}x=\displaystyle\int_{a}^{c}f(x)\,\mathrm{d}x+\displaystyle\int_{c}^{b}f(x)\,\mathrm{d}x.$$ 

Ak niektorý z dvoch nových nevlastných integrálov diverguje, potom hovoríme, že aj pôvodný nevlastný integrál diverguje. **Len** ak obidva nové integrály nadobúdajú konečné hodnoty, potom súčet týchto hodnôt je hodnota pôvodného integrálu.

#### <font color=blue> Príklad 11 </font>   

Vyšetrite konvergenciu integrálu $\displaystyle\int_{-1}^{2}\dfrac{1}{x^4}\,\mathrm{d}x$. 

<ins> Riešenie: </ins> 

Teraz je problematický bod $x=0$ (vertikálna asymptota) a nie krajný bod intervalu $x=-1$ alebo $x=2$. Rozdelíme zadaný nevlastný integrál na dva nové nevlastné integrály, pričom každý z nich bude mať ako krajný bod intervalu (resp. integračnú hranicu) bod $x=0$ (v jeho blízkosti funkcia ubieha do nekonečna). Máme teda 

$$\displaystyle\int_{-1}^{2}\dfrac{1}{x^4}\,\mathrm{d}x=\displaystyle\int_{-1}^{0}\dfrac{1}{x^4}\,\mathrm{d}x+\displaystyle\int_{0}^{2}\dfrac{1}{x^4}\,\mathrm{d}x.$$ 

Na výpočet týchto dvoch integrálov využijeme postup, ktorý už poznáme z predchádzajúcej časti. Zrejme platí 

$$\displaystyle\int_{0}^{2}\dfrac{1}{x^4}\,\mathrm{d}x=\lim\limits_{a\to0^+}\left[-\dfrac{1}{3}x^{-3}\right]_{a}^{2}=\lim\limits_{a\to0^+}\left(-\dfrac{1}{3}\right)\left(\dfrac{1}{8}-\dfrac{1}{a^3}\right),$$

a keďže táto limita nie je konečná, tak integrál $\displaystyle\int_{0}^{2}\dfrac{1}{x^4}\,\mathrm{d}x$ diverguje. Čiže podľa definície uvedenej vyššie aj pôvodný integrál $\displaystyle\int_{-1}^{2}\dfrac{1}{x^4}\,\mathrm{d}x$ diverguje. Podobne sa dá ukázať, že $\displaystyle\int_{-1}^{0}\dfrac{1}{x^4}\,\mathrm{d}x$ diverguje. 

Keď integrand ubieha do nekonečna v blízkosti vnútorného bodu intervalu (nie pri krajnom bode intervalu), ľahko môže človek prehliadnúť, že sa jedná o nevlastný integrál a môže dôjsť k totálne **nesprávnemu výpočtu** zadaného integrálu ako napr. 

$$\displaystyle\int_{-1}^{2}\dfrac{1}{x^4}\,\mathrm{d}x=\left[-\frac{1}{3}x^{-3}\right]_{-1}^{2}=-\frac{1}{24}-\frac{1}{3}=-\frac{3}{8},$$ 

čo je samozrejme **chybné riešenie**. 

```python
x = var('x')
f(x) = 1/x^4
P = line2d([(-1,0),(-1,f(-1))], color='black') 
P += line2d([(2,0),(2,f(2))], color='black') 
P += line2d([(0,0),(0,2.5)], color='orange', linestyle='dashed', thickness=3)
P += polygon([(-1,0),(-1,f(-1))] + [(x, f(x)) for x in [-1,-0.95,..,-0.8]] + [(-0.8,0),(-1,0)],  rgbcolor=(0.8,0.8,0.8),aspect_ratio='automatic')
P += polygon([(0.8,0),(0.8,f(0.8))] + [(x, f(x)) for x in [0.8,0.85,..,2]] + [(2,0),(0.8,0)],  rgbcolor=(0.8,0.8,0.8),aspect_ratio='automatic')
P += polygon([(-0.9,0),(-0.9,2)] + [(x, 2) for x in [-0.9,-0.85,..,0.9]] + [(0.9,0),(-0.9,0)],  rgbcolor=(0.8,0.8,0.8),aspect_ratio='automatic')
P += text("$f(x)=\\dfrac{1}{x^4}$", (2.5, 0.4), fontsize=12, color='blue')
P += plot(f(x), (-3, -0.8), thickness=3) 
P += plot(f(x), (0.8, 3), thickness=3)
P 
```

<font color=orange> Ukážte, že $\displaystyle\int_{-1}^{0}\dfrac{1}{x^4}\,\mathrm{d}x$ z Príkladu 11 diverguje. </font>

#### Poznámka 3 

Ak je integrál $\displaystyle\int_{a}^{b}f(x)\,\mathrm{d}x$ nevlastný (integrand ide do nekonečna) pri oboch krajných bodoch daného intervalu $\langle a,b \rangle$, potom stačí rozdeliť interval v nejakom vnútornom bode $c\in\langle a,b \rangle$ a patrične tak rozdeliť aj pôvodný integrál $\displaystyle\int_{a}^{b}f(x)\,\mathrm{d}x$ na súčet dvoch nových nevlastných  integrálov $\displaystyle\int_{a}^{c}f(x)\,\mathrm{d}x$ a $\displaystyle\int_{c}^{b}f(x)\,\mathrm{d}x$. Pôvodný integrál diverguje ak buď jeden z dvoch nových integrálov diverguje alebo ak obidva divergujú.

#### <font color=blue> Príklad 12 </font>   

Vyšetrite konvergenciu integrálu $\displaystyle\int_{0}^{\infty}\dfrac{1}{x^2}\,\mathrm{d}x$. 

<ins> Riešenie: </ins> 

Zadaný integrál je nevlastný lebo horná integračná hranica je $\infty$, ale je nevlastný tiež kvôli tomu, že integrand $1/x^2$ nie je definovaný v bode $x=0$ a v jeho blízkosti rastie do nekonečna. Rozdelíme teda zadaný integrál napr. v bode $x=1$ na dva nevlastné integrály, konkrétne $\displaystyle\int_{1}^{\infty}\dfrac{1}{x^2}\,\mathrm{d}x$ a $\displaystyle\int_{0}^{1}\dfrac{1}{x^2}\,\mathrm{d}x$. Z Príkladu 5 už vieme, že $\displaystyle\int_{1}^{\infty}\dfrac{1}{x^2}\,\mathrm{d}x$ je konečný. Na druhej strane však $\displaystyle\int_{0}^{1}\dfrac{1}{x^2}\,\mathrm{d}x$ diverguje, lebo 

$$\displaystyle\int_{0}^{1}\dfrac{1}{x^2}\,\mathrm{d}x=\lim\limits_{a\to0^+}\left[-x^{-1}\right]_{a}^{1}=\lim\limits_{a\to0^+}\left(\frac{1}{a}-1\right)=\infty.$$

Preto $\displaystyle\int_{0}^{\infty}\dfrac{1}{x^2}\,\mathrm{d}x$ diverguje.

### Porovnávacie kritérium pre nevlastné integrály 

Niekedy môže byť náročné určiť (nájdením primitívnej funkcie) presne hodnotu nevlastného integrálu, avšak nemusí byť ťažké určiť aspoň to, či daný nevlastný integrál konverguje alebo diverguje, aj to môže byť dôležitá (a postačujúca) informácia v aplikáciách. Podobne ako pri nekonečných číselných radoch, aj tu je niekedy výhodné porovnať daný nevlastný integrál s takým nevlastným integrálom, ktorého správanie poznáme. Ukážme si to na príklade. 

--- 

#### <font color=blue> Príklad 13 </font>   

Vyšetrite konvergenciu $\displaystyle\int_{1}^{\infty}\dfrac{1}{\sqrt{x^3+5}}\,\mathrm{d}x$. 

<ins> Riešenie: </ins> 

Pozrime sa najprv, čo sa deje s integrandom pre $x\to\infty$. Pre veľké $x$ je $5$ zanedbateľná v porovnaní s hodnotami $x^3$, čiže 

$$\dfrac{1}{\sqrt{x^3+5}}\approx\dfrac{1}{\sqrt{x^3}}=\dfrac{1}{x^{3/2}}.$$

Keďže

$$\displaystyle\int_{1}^{\infty}\dfrac{1}{\sqrt{x^3}}\,\mathrm{d}x=\displaystyle\int_{1}^{\infty}\dfrac{1}{x^{3/2}}\,\mathrm{d}x=\lim\limits_{b\to\infty}\displaystyle\int_{1}^{b}\dfrac{1}{x^{3/2}}\,\mathrm{d}x=\lim\limits_{b\to\infty}\left[-2x^{-1/2}\right]_{1}^{b}=\lim\limits_{b\to\infty}\left(2-2b^{-1/2}\right)=2,$$

tak integrál $\displaystyle\int_{1}^{\infty}\dfrac{1}{x^{3/2}}\,\mathrm{d}x$ konverguje. Tým pádom očakávame, že aj pôvodný integrál by mal konvergovať. Aby sme to overili (potvrdili), všimnime si, že platí $0\le x^3\le x^3+5$ a teda máme 

$$\dfrac{1}{\sqrt{x^3+5}}\le\dfrac{1}{\sqrt{x^3}}.$$

Ďalej vďaka poslednej nerovnosti a tiež vďaka vlastnostiam určitého integrálu (nerovnosť medzi funkciami platí aj medzi ich integrálmi) pre $b\ge1$ bude platiť 

$$\displaystyle\int_{1}^{b}\dfrac{1}{\sqrt{x^3+5}}\,\mathrm{d}x\le\displaystyle\int_{1}^{b}\dfrac{1}{\sqrt{x^3}}\,\mathrm{d}x.$$

Teraz by už len stačilo urobiť limitný prechod pre $b\to\infty$ a z vlastností limít vieme, že (neostrá) nerovnosť platiaca pre funkcie sa zachová aj pre limity. Alebo môžeme urobiť nasledujúcu úvahu. 

Aj vďaka nasledujúcemu obrázku si môžeme všimnúť nasledovné: Keďže $\displaystyle\int_{1}^{b}\dfrac{1}{\sqrt{x^3+5}}\,\mathrm{d}x$ sa zväčšuje (žltá plocha), keď $b$ ide do nekonečna ale vždy je to menej (menšia plocha) ako $\displaystyle\int_{1}^{b}\dfrac{1}{\sqrt{x^3}}\,\mathrm{d}x$ a to je určite menej ako $\displaystyle\int_{1}^{\infty}\dfrac{1}{\sqrt{x^3}}\,\mathrm{d}x=2$ (žltá + sivá plocha). Z tejto úvahy teda vieme, že $\displaystyle\int_{1}^{b}\dfrac{1}{\sqrt{x^3+5}}\,\mathrm{d}x$ musí mať konečnú hodnotu menšiu než $2$. Čiže $\displaystyle\int_{1}^{b}\dfrac{1}{\sqrt{x^3+5}}\,\mathrm{d}x$ konverguje k hodnote menšej ako $2$. 


```python
x = var('x')
f(x) = 1/(sqrt(x^3))
g(x) = 1/(sqrt(x^3+5))
P = line2d([(1,0),(1,f(1))], color='black') 
P += polygon([(1,0),(1,f(1))] + [(x, f(x)) for x in [1,1.05,..,4]] + [(4,0),(1,0)],  rgbcolor=(0.8,0.8,0.8),aspect_ratio='automatic')
P += polygon([(1,0),(1,g(1))] + [(x, g(x)) for x in [1,1.05,..,4]] + [(4,0),(1,0)],  color='yellow',aspect_ratio='automatic', alpha=0.4)
P += text("$f(x)=\\frac{1}{\\sqrt{x^3}}$", (1.2, 1.4), fontsize=12, color='blue')
P += text("$g(x)=\\frac{1}{\\sqrt{x^3+5}}$", (3.5, 0.3), fontsize=12, color='red')
P += text("$\\mathrm{Celková\\, plocha (sivá+žltá) }=\\int_{1}^{\\infty}\\frac{1}{\\sqrt{x^3}}\\,\\mathrm{d}x$", (2.1, 1), fontsize=12, color='black')
P += text("$\\mathrm{Žltá\\, plocha }=\\int_{1}^{\\infty}\\frac{1}{\\sqrt{x^3+5}}\\,\\mathrm{d}x$", (2.2, 0.7), fontsize=12, color='black')
P += plot(f, (0.75, 4), thickness=3) 
P += plot(g, (0.75, 4), thickness=3, color='red')
P
```

---

Uvedomme si, že v predošlom príklade sme sa najprv pozreli na správanie sa integrandu pre $x\to\infty$. To je užitočné lebo konvergencia/divergencia integrálu je daná tým, čo sa deje pre $x\to\infty$. 

--- 

**Porovnávacie kritérium pre $\displaystyle\int_{a}^{\infty}f(x)\,\mathrm{d}x$.** 

Nech $f(x)$ je kladná funkcia. Porovnávacie kritérium pozostáva z dvoch krokov: 


1. Podľa správania sa integrandu pre $x\to\infty$ odhadneme, či integrál konverguje alebo nie. 


2. Odhad potvrdíme porovnaním s kladnou funkciou $g(x)$ takou, že platí: 
    * Ak $f(x)\le g(x)$ a $\displaystyle\int_{a}^{\infty}g(x)\,\mathrm{d}x$ konverguje, potom $\displaystyle\int_{a}^{\infty}f(x)\,\mathrm{d}x$ konverguje. 
    * Ak $g(x)\le f(x)$ a $\displaystyle\int_{a}^{\infty}g(x)\,\mathrm{d}x$ diverguje, potom $\displaystyle\int_{a}^{\infty}f(x)\,\mathrm{d}x$ diverguje.

#### <font color=blue> Príklad 14 </font>   

Vyšetrite konvergenciu $\displaystyle\int_{4}^{\infty}\dfrac{1}{(\ln t)-1}\,\mathrm{d}t$. 

<ins> Riešenie: </ins>  

Keďže $\ln t$ rastie bez obmedzenia pre $t\to\infty$, potom $-1$ je zanedbateľné v porovnaní s $\ln t$. Preto keď uvažujeme o konvergencii, tak je potrebné si uvedomiť, že $\displaystyle\int_{4}^{\infty}\dfrac{1}{(\ln t)-1}\,\mathrm{d}t$ sa správa ako $\displaystyle\int_{4}^{\infty}\dfrac{1}{\ln t}\,\mathrm{d}t$. Otázka teraz znie, či $\displaystyle\int_{4}^{\infty}\dfrac{1}{\ln t}\,\mathrm{d}t$ konverguje alebo diverguje? Intuitívne môžeme vytušiť, že $\ln t$ rastie veľmi pomaly a teda $1/\ln t$ ide k nule veľmi pomaly, čo môže spôsobiť divergenciu integrálu. 

Zrejme platí $(\ln t)-1<\ln t<t$ pre všetky $t>0$. Za predpokladu $t>e$ môžeme vziať prevrátené hodnoty a dostaneme 

$$\dfrac{1}{(\ln t)-1}>\dfrac{1}{\ln t}>\dfrac{1}{t}.$$ 

Keďže $\displaystyle\int_{4}^{\infty}\dfrac{1}{t}\,\mathrm{d}t$ diverguje (prečo?), vzhľadom k nerovnostiam uvedeným vyššie môžeme urobiť záver, že aj $\displaystyle\int_{4}^{\infty}\dfrac{1}{(\ln t)-1}\,\mathrm{d}t$ diverguje.
