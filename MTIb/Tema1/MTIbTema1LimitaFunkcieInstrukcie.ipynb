# <font color=brown> Motivácia </font>

V informatike **pri skúmaní algoritmov** je okrem iného zaujímavá aj tzv. **asymptotika v rámci časovej i pamäťovej zložitosti algoritmov**. Skúma sa, ako dlho pobeží daný algoritmus alebo koľko pamäte zaberie, keď sa vstup algoritmu zväčšuje a rastie až k extrémnym hodnotám (teoreticky až do nekonečna). Na základe tohto (limitného) správania sa algoritmu (v "najhoršom" možnom/očakávanom prípade/scenári), je možné algoritmy porovnávať, odhadovať ich výpočtový čas či potrebnú pamäť a následne tieto algoritmy optimalizovať alebo paralelizovať atď. 

Ak by sme **modelovali rozlíšenie obrazovky** (počet pixelov - malé štvorčeky vypĺňajúce obrazovku) ako nejakú funkciu a vzali by sme jej **limitu**, keď sa rozlíšenie blíži k $0$, dostali by sme reálne číslo. Toto číslo je rovnaké rozlíšenie, aké používajú počítačové tlačiarne, t.j. koľko „malých bodiek“ má byť umiestnených na štvorcový palec, aby sa to javilo ako pekná (hladká) fotografia. Ak by sme vzali **limitu**, keď sa zmena vzdialenosti na výtlačku blíži k $0$, nakoniec by sme videli bodky. To je tiež dôvod, prečo HD televízory prichádzajú s rozlíšením 1080p, a nie s nekonečne veľa pixelmi. Pri rozlíšení 1080p je rozdiel v rozlíšení toho, čo naše oči môžu prijať od toho, čo dokáže vyprodukovať televízor, dostatočne blízko k $0$ na to, aby na tom už v praxi nezáležalo. Rovnako ako keď by sme povedali: „$0.000000000001$ je dosť blízko k $0$, aby to mohla byť aj $0$“, pre výrobcu TV platí rovnaká logika. To šetrí náklady na pixeloch na televízore a pre spotrebiteľa to bude v bežnom živote rovnaký výsledok (efekt).

 

Podobné je to s **audio súbormi. MP3** je súbor diskrétnych zvukových bajtov, ktoré sa prehrávajú v sekvencii, ale za sekundu ich hrá toľko, že to znie ako "hladký" súvislý zvuk resp. hudba vychádzajúca z reproduktora. Toto je ďalší technický problém, pri ktorom vstupujú do hry **limity**. Aká je zmena času potrebná na to, aby sa jednotlivé zvukové bajty v ľudskom uchu javili ako plynulé? Vezme sa **limita matematického modelu**, zistí sa, k čomu sa blíži, a na základe toho sa nastaví intezita (tempo) prehrávaných zvukových bajtov tak, aby zneli dobre. 

 

Okrem už spomínaných aplikácii **limít funkcií**, sú **limity** dôležité a sú základom pre také pojmy ako spojitosť a derivácia, s ktorými sa stretneme neskôr kde uvidíme ich silu a využitie. 

Náplňou témy 1 je teda pojem **limita funkcie**. Je to fundamentálny a veľmi dôležitý koncept, preto je nutné ho správne uchopiť a pochopiť. Pozrieme sa na niekoľko typov **limít - vlastné, nevlastné, limity v nevlastných bodoch, či jednostranné limity**. Okrem samotných definícii si tiež názorne vysvetlíme **geometrickú interpretáciu** rôznych typov **limít**. Vyslovíme niekoľko matematických viet pre **výpočet** základných typov **limít** a budeme ich ilustrovať na konkrétnych príkladoch.
