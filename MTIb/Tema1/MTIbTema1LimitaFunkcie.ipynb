#### <font color=blue> Príklad (okamžitá rýchlosť padajúcej gule - úvod) </font> 

Uvažujme funkciu $s(t)$, ktorá vyjadruje polohu pohybujúceho sa objektu v čase $t$. Konkrétne nech $s(t)=64-16t^2$ je poloha padajúcej gule v čase $t$. 

```python
t =  var('t')
s(t) = 64-16*t^2 
kwds = {'fontsize':12} 
plot(s(t), (t, 0, 2), gridlines = 'minor', axes_labels=['$x$', '$y$'], axes_labels_size = 1.2, figsize=4) + text(r"$s(t)=64-16t^2$", (1.5, 50), color='blue',**kwds) 
```

Funkcie môžu modelovať ďalšie zaujímavé javy, ako napríklad rýchlosť, akou automobil spotrebúva benzín pri danej rýchlosti, alebo reakcia pacienta na danú dávku lieku. Pomocou nástrojov kalkulu môžeme študovať, ako sa funkčná hodnota mení v závislosti od zmeny vstupnej (nezávislej) premennej. 

Ak sa vrátime k funkcii popisujúcej polohu padajúcej gule v čase $t$, tak priemerná rýchlosť ($AV_{\langle 1,x \rangle}$ - Average Velocity) padajúcej gule na intervale $\langle 1,x \rangle$ je daná podielom 

$$AV_{\langle 1,x \rangle}=\dfrac{s(x)-s(1)}{x-1}=\dfrac{(64-16x^2)-(64-16\cdot 1)}{x-1}=\dfrac{16-16x^2}{x-1},$$ 

t.j. celková dráha deleno celkový čas. 

Všimnime si, že priemerná rýchlosť je funkciou premennej $x$. Ináč povedané, funkcia $h(x)=\dfrac{16-16x^2}{x-1}$ udáva priemernú rýchlosť gule na intervale od $t=1$ po $t=x$. Aby sme našli okamžitú rýchlosť (tá sa prakticky nedá namerať priamo) gule, keď $t = 1$, musíme vedieť, čo sa stane s $h(x)$, keď sa $x$ dostáva bližšie a bližšie k $1$. Ale tiež si všimnime, že $h(1)$ nie je definovaná, pretože vedie k podielu $0/0$. 

Tu prichádza na scénu pojem _limity_. Použitím limity môžeme skúmať správanie funkcie $h(x)$, keď sa $x$ približuje ľubovoľne blízko k $1$ ale nerovná sa $1$. Najprv je vhodné pomocou grafu funkcie preskúmať body, v ktorých dochádza k zaujímavému správaniu sa  funkcie.

**Limity nám umožňujú identifikovať trend v hodnotách funkcie, keď sa jej vstupná (nezávislá) premenná približuje k nejakej konkrétnej hodnote.** Potrebujeme presné pochopenie toho, čo to znamená povedať, „funkcia $f$ má limitu $A$, keď sa $x$ blíži k $x_0$.“ Inými slovami, potrebujeme presne vymedziť to, že "$f(x)$ je blízko $A$ keď $x$ je blízko $x_0$" resp., že "s $f(x)$ sa vieme dostať ľubovoľne blízko k hodnote $A$ keď sme s $x$ dostatočne blízko k $x_0$". Presne toto nám formálne a exaktne vystihuje definícia limity uvedená ďalej, avšak ešte pred tým sa vráťme na chvíľu k príkladu funkcie znázornenej na poslednom obrázku. 

V predchádzajúcom príklade sme videli (pozri graf), že keď sa $x$ približuje (sprava alebo zľava) stále viac a viac k $0$  (ale nerovná sa $0$), potom hodnota funkcie $g$ sa priblíži tak blízko k hodnote $4$, ako len chceme (ľubovoľne blízko). Na prvý pohľad sa to môže zdať kontraintuitívne (odporujúce našej predstave), pretože hodnota $g(0)$ je $1$, nie $4$. Limity však opisujú správanie funkcie v ľubovoľnej blízkosti pevného vstupu (v tesnej blízkosti nejakého $x_0$) a na hodnote funkcie vo zvolenom pevnom vstupe nezáleží (hodnota $f(x_0)$ nie je podstatná pokiaľ sa jedná o limitu v bode $x_0$). Preto má zmysel definovať niečo ako _prstencové okolie bodu_ $x_0$. 

Zavedieme ešte pomocné pojmy a následne vyslovíme definíciu limity funkcie v bode, kde formálne zhrnieme všetky predchádzajúce úvahy.

### <font color=red> Definícia 1 </font> 

* **_Okolie bodu $x_0$_** ozn. $\mathcal{O}(x_0)$: ľubovoľný otvorený interval obsahujúci bod $x_0$; 


* **_$\delta$-okolie bodu $x_0$_** ozn. $\mathcal{O}_{\delta}(x_0)$: množina takých bodov $x$ na reálnej osi, ktoré sú vzdialené od bodu $x_0$ o menej ako $\delta$ t.j. $\mathcal{O}_{\delta}(x_0)=\{x\in\mathbb{R}:\,|x-x_0|<\delta\}$, kde $\delta>0$; 


* **_Prstencové okolie bodu $x_0$_** ozn. $\mathcal{O}^{*}(x_0)$: okolie bodu $x_0$ bez samotného bodu $x_0$ t.j. $\mathcal{O}^{*}(x_0)=\mathcal{O}(x_0) \backslash \{x_0\}$; 


* **_Prstencové $\delta$-okolie bodu $x_0$_** ozn. $\mathcal{O}^{*}_{\delta}(x_0)$: delta okolie bodu $x_0$ bez samotného bodu $x_0$ t.j. $\mathcal{O}^{*}_{\delta}(x_0)= \mathcal{O}_{\delta}(x_0) \backslash \{x_0\} =\{x\in\mathbb{R}:\,0<|x-x_0|<\delta\}$, kde $\delta>0$; 


* **_Okolie bodu $+\infty$_** ozn. $\mathcal{O}(+\infty)$: ľubovoľný otvorený interval $(a,+\infty)$, kde $a\in\mathbb{R}$; 


* **_Okolie bodu $-\infty$_** ozn. $\mathcal{O}(-\infty)$: ľubovoľný otvorený interval $(-\infty,b)$, kde $b\in\mathbb{R}$.

#### Poznámka 1 

* Zrejme funkcia $\sin\,x$ je definovaná na celej reálnej osi tzn. $\forall x\in\mathbb{R}$. Naproti tomu funkcia $\dfrac{\sin\,x}{x}$ nie je definovaná (pozri graf nižšie, prázdny krúžok v $[0,1]$) v bode $x_0=0$ (nula v menovateli zlomku je neprípustná), avšak je definovaná pre každé $x$, ktoré je ľubovoľne blízko k $x_0$ a zároveň je rôzne od $x_0$. Inými slovami to znamená, že funkcia $\dfrac{\sin\,x}{x}$ je definovaná na nejakom prstencovom okolí bodu $x_0$, pričom za prstencové okolie bodu $x_0=0$ funkcie $\dfrac{\sin\,x}{x}$ môžeme vziať napr. interval $(-0.1, 0.1)$ okrem bodu $0$ t.j. $(-0.1, 0.1)\backslash \{0\}$ ale napr. aj $(-0.05, 0.05)\backslash \{0\}$.  # <font color=brown> Vlastná limita (vo vlastnom bode)</font>

---

### <font color=red> Definícia 2 </font> 

Nech funkcia $f:\ \mathbb{R} \rightarrow \mathbb{R}$ je definovaná na nejakom okolí bodu $x_0$ (pričom nemusí byť definovaná v $x_0$, tzn. stačí, keď je funkcia definovaná na nejakom prstencovom okolí bodu $x_0$). Hovoríme, že: 
		
funkcia $f$ má v bode $x_0$ **_limitu_** rovnú $A$, ak

$$\left(\forall \varepsilon > 0\right)\left(\exists \delta > 0\right)\left(\forall x \in (x_0 - \delta,x_0+\delta),x\neq x_0\right)\ |f(x)-A| < \varepsilon,$$

funkcia $f$ má v bode $x_0$ **_sprava_** limitu rovnú $A$, ak

$$(\forall \varepsilon > 0)(\exists \delta > 0)(\forall x \in (x_0,x_0+\delta))\ |f(x)-A| < \varepsilon,$$

funkcia $f$ má v bode $x_0$ **_zľava_** limitu rovnú $A$, ak

$$(\forall \varepsilon > 0)(\exists \delta > 0)(\forall x \in (x_0 - \delta,x_0))\ |f(x)-A| < \varepsilon.$$

Limitu $f$ v bode $x_0$ (resp. v bode $x_0$ sprava alebo zľava) označujeme $\lim\limits_{x \to x_0} f(x)$ (resp. $\lim\limits_{x \to x_0+} f(x)$, $\lim\limits_{x \to x_0-} f(x)$).
	
    
Takto definovaná limita sa zvykne tiež nazývať **_vlastná limita vo vlastnom bode_**. 

---

---

#### Poznámka 2 

* $x$ sa blíži k $x_0$ zvykneme označovať $x \rightarrow x_0$, $x$ sa blíži k $x_0$ sprava zvykneme označovať $x \rightarrow x_0^+$ a napokon $x$ sa blíži k $x_0$ zľava zvykneme označovať $x \rightarrow x_0^-$. 


* Definícia limity funkcie v bode nám vlastne hovorí, že s hodnotou $f(x)$ sa vieme priblížiť ľubovoľne blízko k hodnote $A$ (tomu odpovedá v definícii limity $\forall\, \varepsilon > 0:\,|f(x)-A| < \varepsilon$), ak sa s $x$ priblížime dostatočne blízko k $x_0$ tzn. sme v nejakom (malom) vhodnom prstencovom okolí bodu $x_0$ (tomu odpovedá v definícii limity $\exists\, \delta > 0$ také, že $\forall x \in (x_0 - \delta, x_0 + \delta),x\neq x_0$). Všimnime si, že $\varepsilon$ nám udáva akúsi toleranciu resp. maximálnu vzdialenosť $f(x)$ od $A$ na nejakom malom okolí bodu $x_0$. Inými slovami, ak funkcia $f$ má limitu $A$ v bode $x_0$, tak nech zvolíme $\varepsilon>0$ akokoľvek malé, vždy je možné nájsť k nemu také $\delta$, že pre $x$ z tohto prstencového delta okolia bodu $x_0$ bude platiť $|f(x)-A|<\varepsilon$, tzn. zmenšovaním $\varepsilon$ sa teda vieme dostať s $f(x)$ ľubovoľne blízko ($|f(x)-A|$ vyjadruje vzdialenosť $f(x)$ od $A$) k hodnote $A$ na nejakom okolí bodu $x_0$. 



* Limita sprava hovorí o približovaní sa k $x_0$ po reálnej osi len sprava a teda pozeráme sa na tzv. _pravé prstencové okolie_ bodu $x_0$ t.j. $(x_0,x_0+\delta)$. Podobne limita zľava hovorí o približovaní sa k $x_0$ po reálnej osi iba zľava a teda pozeráme sa na tzv. _ľavé prstencové okolie_ bodu $x_0$ t.j. $(x_0-\delta,x_0)$. 


* Jednostranná limita v bode $x_0$ teda slúži na skúmanie správania sa funkcie buď napravo alebo naľavo od bodu $x_0$ t.j. skúmame správanie sa funkcie keď sa k $x_0$ približujeme buď sprava alebo zľava. Limita funkcie v bode $x_0$ v sebe zahŕňa obidva tieto prípady. 


* Ešte je dôležité si uvedomiť, že **v definícii limity, to $\delta$, ktoré tam vystupuje, závisí jednak od zvoleného $\varepsilon$ ale tiež od bodu $x_0$, v ktorom limitu hľadáme**. Ak by sme chceli byť úplne striktní a dôslední, mohli by sme písať, že $\delta (x_0, \varepsilon)$, nezvykne sa to však robiť, no majme to na pamäti.    


* Z vlastností absolútnej hodnoty vyplýva, že výraz $|f(x)-A|<\varepsilon$ v definícii limity môžeme ekvivalentne prepísať takto: $A-\varepsilon < f(x) < A+\varepsilon$. Toto vyjadrenie nám vlastne hovorí o páse šírky $2\varepsilon$ okolo hodnoty $A$ ($\varepsilon$-vzdialenosť smerom nahor od $A$, $\varepsilon$-vzdialenosť smerom nadol od $A$), v ktorom ležia funkčné hodnoty $f(x)$, ak $x$ je z dostatočne malého okolia bodu $x_0$. Tvar $A-\varepsilon < f(x) < A+\varepsilon$ nás priamo navádza ku geometrickej interpretácii limity (pozri obrázok a interaktívny kód nižšie). 


* Podmienku $\forall x \in (x_0 - \delta,x_0+\delta),x\neq x_0$ v Definícii 2 je možné ekvivalentne vyjadriť takto: $\forall x\in\mathcal{D}(f):0<|x-x_0|<\delta$.

#### <font color=blue> Príklad 1 </font> 

Vráťme sa k príkladu s funkciou $g(x)$, ktorá je znázornená na obrázku. 


$$\text{Zdroj: Boelkins (2018).}$$

Z grafu môžeme ľahko vyčítať, že platí 

$$\lim\limits_{x \to -1} g(x)=3, \qquad\qquad \lim\limits_{x \to 0} g(x)=4, \qquad\qquad \lim\limits_{x \to 2} g(x)=1.$$

Keď sa snažíme určiť limitu funkcie pomocou jej grafu, stačí overiť, či sa funkčné hodnoty $f(x)$ blížia k jednej a tej istej hodnote $A$ sprava aj zľava keď sa $x$ blíži k (fixnému) $x_0$ sprava aj zľava. Keď sa blížime k $x_0=1$ zľava tak funkčné hodnoty $g(x)$ sa blížia k hodnote $3$ ale keď sa blížime k $x_0=1$ sprava tak funkčné hodnoty $f(x)$ sa blížia k hodnote $2$. Neexistuje jedno a to isté číslo, ku ktorému by sa tieto funkčné hodnoty blížili (keď sa približujeme k $x_0=1$). To je dôvod, prečo funkcia $g$ nemá v bode $x_0=1$ limitu.

Z výpisu hodnôt sa zdá, že s hodnotu $f$ sa vieme dostať ľubovoľne blízko k hodnote $3$ ak vezmeme $x$ dostatočne blízko k $-1$ a to nás nabáda k tomu, že $\lim\limits_{x \to -1} f(x)=\lim\limits_{x \to -1} \dfrac{4-x^2}{x+2}=3$. Tento záver je v zhode s grafom funkcie $f$. Pozrime sa na to trochu exaktnejšie a z algebraického (resp. analytického) pohľadu, teda využijeme predpis funkcie $f(x)=\dfrac{4-x^2}{x+2}$. Keď $x\rightarrow-1$, potom $(4-x^2)\rightarrow\left(4-(-1)^2\right)=3$ a $(x+2)\rightarrow(-1+2)=1$, čiže ak $x\rightarrow-1$, potom hodnota čitateľa zlomku v predpise funkcie $f$ sa blíži k $3$ a hodnota menovateľa zlomku v predpise funkcie $f$ sa blíži k $-1$, preto $\lim\limits_{x \to -1} f(x)=\lim\limits_{x \to -1} \dfrac{3}{1}=3$. 

Trochu zložitejšia je situácia keď $x\rightarrow-2$ lebo $f(-2)$ nie je definované (nula v menovateli zlomku). Ak by sme skúsili analogicý algebraický prístup ako v predošlom prípade, tak pre $x\rightarrow-2$ máme $(4-x^2)\rightarrow(4-(-2)^2)=0$ a $(x+2)\rightarrow(-2+2)=0$, čiže keď $x\rightarrow-2$, potom čitateľ aj menovateľ zlomku v predpise funkcie $f$ sa blížia k $0$, čím vzniká tzv. _nedefinovaný výraz_ v tvare $0/0$. To naznačuje, že treba urobiť ešte čosi iné, coši viac, aby sme vypočítali limitu v bode $-2$, ak taká limita existuje. Na druhej strane z grafu funkcie $f$ a z výpisu funkčných hodnôt (viď hore) v blízkom okolí $2$ sa zdá, že funkcia $f$ by mala mať v bode $2$ limitu rovnú hodnote $4$. <br/>
Pozrime sa na to algebraicky ale trochu ináč. Keď skúmame limitu funkcie $f(x)=\dfrac{4-x^2}{x+2}$ v bode $-2$, tak nás vlastne zaujíma správanie sa tejto funkcie v (ľubovoľnej) tesnej blízkosti bodu $-2$, nikdy nie však priamo v bode $-2$. Všimnime si, že keď vo výraze $\dfrac{4-x^2}{x+2}=\dfrac{(2-x)(2+x)}{x+2}$ vykrátime výraz $(x+2)$ (ktorý je v súčine v čitateli zlomku a tiež v menovateli zlomku), tak dostaneme výraz $2-x$. Funkcia $f(x)=\dfrac{4-x^2}{x+2}=\dfrac{(2-x)(2+x)}{x+2}$ a funkcia $2-x$ sa rovnajú na celej reálnej osi s výnimkou bodu $-2$, keďže tam funkcia $f$ definovaná nie je, zatiaľ čo funkcia $2-x$ áno. Pri skúmaní limity funkcie $f$ v bode $-2$ ale funkčná hodnota v samotnom bode $-2$ nie je dôležitá, pohybujeme sa totiž len na prstencovom okolí (podľa definíce limity) bodu $-2$ (t.j. okrem bodu $-2$), preto pri výpočte tejto limity môžeme miesto funkcie $f$ uvažovať ďalej funkciu $2-x$, ktorá vznikla po vykrátení. Na základe týchto úvah môžeme písať 

$$\lim\limits_{x \to -2} f(x)=\lim\limits_{x \to -1} (2-x)=4.$$

Túto poslednú uvedenú limitu už ľahko určíme (či už z grafu funkcie $2-x$ alebo cez postupnosť jej funkčných hodnôt v blízkosti bodu $-2$) a teda $\lim\limits_{x \to -2} f(x)=4$.

Presná hodnota $g(3)=\sin\left(\dfrac{\pi}{3}\right)$ je $\dfrac{\sqrt{3}}{2}$ (pozri jednotkovú kružnicu, graf alebo tabuľky), čo je približne $0.8660254038$. To nás vedie k záveru, že 

$$\lim\limits_{x \to 3} g(x) = \lim\limits_{x \to 3} \sin\left(\dfrac{\pi}{3}\right) = \dfrac{\sqrt{3}}{2}. $$ 

Všimnime si, že keď $x \rightarrow 0$ tak výraz $\dfrac{\pi}{x}$ sa nespráva celkom "štandardne". Keď je $x>0$ a blíži sa k $0$ (sprava), tak delíme číslo $\pi$ stále menším a menším kladným číslom (blízkym nule), čo spôsobuje, že hodnota zlomku $\dfrac{\pi}{x}$ rastie nad všetky hodnoty až do $+\infty$. Na druhej strane, keď je $x<0$ a blíži sa k $0$ (zľava) tak hodnota zlomku klesá pod všetky medze až do $-\infty$. V tomto zmysle teda keď sa $x$ blíži k $0$ tak vstupy v tvare $\dfrac{\pi}{x}$ pre funkciu sínus sa rapídne zväčšujú (resp. zmenšujú) a to vedie k rapídne sa zväčšujúcim osciláciám grafu funkcie $g(x)=\sin\,\left(\dfrac{\pi}{x}\right)$ medzi hodnotami $-1$ a $1$. Osciláciami rozumieme "divoké" vlnenie sa grafu funkcie (pripomínajúce akoby rýchle kmity medzi) medzi $-1$ a $1$ v okolí bodu $0$ (viď graf hore). Pri vykreslení grafu funkcie $g$ na malom okolí bodu $0$ (vykreslite si!) je vidieť, že sa funkčná hodnota nikdy neustáli okolo nejakého jedného čísla v blízkosti počiatku súradnicovej sústavy. Toto nás vedie k záveru, že funkcia $g$ nemá v bode $x_0=0$ limitu. 

Ako ale zladiť to, čo napovedá graf funkcie $g$ s výpisom funkčných hodnôt v okolí nuly (pozri hore), ktorý naznačuje, že limita $g$ pre $x \rightarrow 0$ by mohla byť $0$ ? Výpis funkčných hodnôt je v tomto prípade zavádzajúci kvôli špeciálnemu tvaru (povahe) zvolenej postupnosti hodnôt $\{0.1,0.001,0.0001,\ldots\}$ v blízkosti nuly. Členy tejto postunposti sú vlastne čísla v tvare $10^{-k}$, kde $k$ ide cez prirodzené čísla. Keď vypočítame $g(10^{-k})=\sin\,\left(\dfrac{\pi}{10^{-k}}\right)=\sin\,(10^{k}\pi)=0$ pre každé $k\in\mathbb{N}$ (pozri jednotkovú kružnicu, graf alebo tabuľky). Keď ale vezmeme inú postupnosť čísel (bodov) v blízkosti nuly, napr. postupnosť $\{0.3,0.03,0.003,\ldots\}$, čo sú vlastne čísla v tvare $3\cdot10^{-k},\,k\in\mathbb{N}$, tak zistíme, že 

$$g(3\cdot10^{-k})=\sin\,\left(\dfrac{\pi}{3\cdot10^{-k}}\right)=\sin\,\left(\dfrac{10^k\pi}{3}\right)=\dfrac{\sqrt{3}}{2}\approx 0.866025.$$

Táto postupnosť funkčných hodnôt nám na rozdiel od predošlej postupnosti podsúva myšlienku, že by limita funkcie $g$ v bode $0$ mohla byť $\dfrac{\sqrt{3}}{2}$. **Podľa vety o jednoznačnosti limity (Veta 2) ale vieme, že to nie je možné aby funkcia mala v jednom bode dve limity zodpovedajúce rôznym číslam.** Preto funkcia $g$ nemá limitu pre $x \rightarrow 0$. 

Výpis funkčných hodnôt v okolí nejakého bodu môže byť užitočný na preskúmanie správania funkcie v okolí daného bodu, no nie je možné sa na neho na 100% spoľahnúť a určiť limitu iba na jeho základe. Ako sme videli v tomto príklade, pri určení/výpočte limity je potrebné využiť aj iné nástroje (prípadne ich skombinovať). 

Na záver tohto príkladu ešte kontrolný výpočet limít funkcie $g$ pomocou Sage.

#### <font color=blue> Príklad 3 (okamžitá rýchosť padajúcej gule - dokončenie) </font> 

V úvode tejto témy sme videli, že priemernú rýchlosť padajúcej gule v čase od $1$ po $x$ časových jednotiek t.j. na intervale $\langle 1, x  \rangle$ môžeme vyjadriť funkciou 

$$AV_{\langle 1,x \rangle}=h(x)=\dfrac{16-16x^2}{x-1}.$$

Okamžitá rýchlosť padajúcej gule v čase $t=1$ ozn. $IV_{t=1}$ sa definuje ako limita funkcie $h(x)$ udávajúcej priemernú rýchlosť keď sa $x$ blíži k $1$. Predstavme si postupnosť takých $x$, ktoré sa blížia k $1$ a tomu odpovedá postupnosť funkčných hodnôt $h(x)$, ktorá sa približuje k okamžitej rýchlosti v čase $t=1$. Všimnime si, že keď $x \rightarrow 1$ tak časový interval na ktorom uvažujeme priemernú rýchlosť $h(x)$ sa skracuje. Na základe týchto úvah pre okamžitú rýchlosť padajúcej gule v čase $t=1$ platí 

$$IV_{t=1}=\lim\limits_{x \to 1} AV_{\langle 1,x \rangle} = \lim\limits_{x \to 1} \dfrac{s(x)-s(1)}{x-1} = \lim\limits_{x \to 1} h(x) = \lim\limits_{x \to 1} \dfrac{16-16x^2}{x-1}=\lim\limits_{x \to 1} \dfrac{-16(x^2-1)}{x-1} = \lim\limits_{x \to 1} \dfrac{-16(x-1)(x+1)}{x-1} = \lim\limits_{x \to 1} -16(x+1)=-32.$$

Okamžitá rýchlosť padajúcej gule v čase $t=1$ je $32$ (napr. m/s alebo nejaká iná jednotka). 
Záporné znamienko (rýchlosť je vektor tzn. má smer aj veľkosť, v tomto prípade je to jednorozmerný vektor) len hovorí o tom, že guľa padá smerom nadol (pohybuje sa v opačnom resp. "zápornom" smere). 
Všimnime si ešte, že na výpočet tejto limity sme použili podobnú úvahu i algebraickú úpravu ako v Príklade 2.  
Kľúčovou myšlienkou je to, že na výpočet okamžitej rýchlosti berieme limitu priemerných rýchlostí, keď sa časový interval skracuje. Tento typ limity je v praxi a v aplikáciách veľmi dôležitý, preto má aj samostatné pomenovanie (derivácia) a bude mu venovaná  jedna z ďalších tém v rámci tohto kurzu. 

---

**Geometrický význam vlastnej limity v bode**: pre každú šírku pásu (rovnú $2\varepsilon$, kde $\varepsilon$ je vzdialenosť okraja pásu od jeho čiarkovanej osi) s horizontálnou osou $y = A$ existuje okolo bodu $x_0$ interval polomeru $\delta$ taký, že úsek grafu funkcie (červená krivka) vymedzený krajnými bodmi tohto intervalu leží celý vnútri daného pásu. Toto musí platiť pre každy taký pás šírky $2\varepsilon$ okolo priamky $y=A$, ak $A$ je limita danej funkcie v bode $x_0$. To isté vlastne hovorí **definícia limity funkcie v bode** t.j. $\left(\forall \varepsilon > 0\right)\left(\exists \delta > 0\right)\left(\forall x \in (x_0 - \delta,x_0+\delta),x\neq x_0\right)\ |f(x)-A| < \varepsilon$ resp. **ekvivalentné vyjadrenie** $\left(\forall \varepsilon > 0\right)\left(\exists \delta > 0\right)\left(\forall x\in\mathcal{D}(f):0<|x-x_0|<\delta\right)\ |f(x)-A| < \varepsilon$. 


$$\text{Zdroj: Madaras (2019).}$$

#### Poznámka 4 

V prípade jednostranných limít t.j. limity sprava resp. limity zľava by nás zaujímali iba tie $x$, ktoré sú napravo od $x_0$ resp. naľavo od $x_0$, teda tie, ktoré sú z tzv. **_pravého okolia_** bodu $x_0$ t.j. $(x_0,x_0+\delta)$ resp. z **_ľavého okolia_** bodu $x_0$ t.j. $(x_0-\delta,x_0)$. Pre obrázok vyššie by to znamenalo, že zo sivého pásu by sme odrezali časť naľavo od $x_0$ resp. napravo od $x_0$. 

---

**Definíciu limity a jej geometrickú interpretáciu si môžete výskúšať spustením nasledovného interaktívneho kódu**, ktorý má tieto vstupné parametre: $f$ - funkcia, ktorej limitu skúmame; $x_{min}$ a $x_{max}$ - určujú rozsah na osi $x$, kde sa má vykresliť graf funkcie $f$; $x_0$ - bod, v ktorom skúmame (hľadáme) limitu funkcie $f$; $A$ - náš tip, čo by mohla byť limita funkcie $f$ v bode $x_0$; $\varepsilon$ - polovica šírky priehľadného horizontálneho pásu okolo hodnoty $A$, ohraničeného dvoma prerušovanými čiarami (t.j. okolie bodu $y=A$); $\delta$ - polovica šírky ružového vertikálneho pásu okolo bodu $x_0$, ohraničeného dvoma prerušovanými čiarami (t.j. okolie bodu $x_0$). 

**Princíp experimentovania je nasledovný:** Zadajte nejakú funkciu $f$; nastavte rozumný interval pomocou $x_{min}$ a $x_{max}$, na ktorom sa má vykresliť graf funkcie $f$; zvoľte bod $x_0$, v ktorom chcete skúmať (hľadať) limitu danej funkcie $f$; potom odhadnite, aká by podľa Vás mohla byť limita funkcie $f$ v bode $x_0$ tzn. zadajte hodnotu $A$; ďalej zvoľte nejaké (malé) $\varepsilon>0$ a následne skúšajte k zvolenému $\varepsilon$ nájsť (malú) hodnotu $\delta>0$ (či existuje) takú, aby graf funkcie $f$ nezasahoval do ružového pásu (ale aby bol "stlačený" v priesvitnom $\varepsilon$-páse). Ak funkcia $f$ má v bode $x_0$ limitu $A$, potom ku každému $\varepsilon>0$ by ste mali vedieť nájsť také vhodné $\delta>0$. To je len inými slovami prerozprávaná definícia limity. Samozrejme, nie je prakticky možné vyskúšať ku každému $\varepsilon>0$ (je ich nekonečne veľa) nájsť vhodné $\delta>0$ (overiť definíciu limity na 100%) ale podstatné je porozumieť, čo nám definícia limity hovorí, ako ju vieme experimentálne overovať (aj keď nie na 100%) a ako to vyzerá geometricky.

Z definície limity dokážte, že $\lim\limits_{x \to 4} \sqrt{x} = 2$.

<ins> Riešenie: <ins/> 

Pred tým, než použijeme formálnu definíciu, urobme niekoľko konkrétnych numerických experimentov, aby sme lepšie porozumeli, o čo ide. Vezmime najprv $\varepsilon=0.5$, ako blízko (aké má byť delta?) musí byť $x$ k bodu $x_0=2$, aby $2-\varepsilon <  f(x) < 2+\varepsilon$ t.j. $2-0.5 <  f(x) < 2+0.5$ t.j. $1.5 <  f(x) < 2.5$ ? Ináč povedané, pre aké $\delta$, ak zoberieme ľubovoľné $x\in(4-\delta,4+\delta)\backslash\{2\}$ bude platiť $1.5 <  f(x) < 2.5$ ? Môžeme to zistiť nasledovne: 

\begin{align*}
1.5 <\,  & f(x) < 2.5 \\ 
1.5 <\, & \sqrt(x) < 2.5 \\ 
(1.5)^2 <\, & x < (2.5)^2 \\ 
2.25 <\, & x < 6.25 
\end{align*}

Aké má byť teda $\delta$ pre $\varepsilon=0.5$ ? Hľadáme $\delta$ tak aby sme dostali symetrický interval so stredom v $x_0=4$ t.j. $4-\delta < x < 4+\delta$ resp. $(4-\delta,4+\delta)$. Z nerovností uvedených vyššie máme, že musí platiť $2.25<x$ a tomu  odpovedá $4-\delta < x$, odtiaľ $4-\delta>2.25$, čo dáva $\delta<1.75$. Na druhej strane z nerovností uvedených vyššie tiež máme, že musí platiť $x<6.25$ a tomu odpovedá $x<4+\delta$, čo dáva spolu $6.25>4+\delta$ a teda $\delta<2.25$. V našom prípade potrebujeme vybrať len jedno $\delta$ a konkrétne vyberieme $\delta$ menšie ako minimum z dvoch ponúkajúcich sa hodnôt, t.j. $\delta<\min\{1.75,2.25\}=1.75$. Minimum berieme preto, aby boli dodržané hranice pre $x$, ktoré sme vypočítali vyššie, konkrétne $2.25 < x < 6.25 $. Toto zrejme bude splnené, keď vezmeme $\delta<1.75$ a potom máme $2.25<x<5.75$. Inými slovami, ak vezmeme ľubovoľné $x\in(2.25,5.75)\backslash\{4\}$, potom pre každé také $x$ bude platiť $1.5 <  f(x) < 2.5$. <font color=orange> Overte to pomocou interaktívneho kódu ! Taktiež skúste v interaktívnom kóde zobrať $\delta$ ako maximum z čísel $1.75$ a $2.25$ miesto minima a všimnite si, prečo také $\delta$ nie je vhodné. </font>

Vezmime teraz $\varepsilon=0.01$. Aké by malo byť v tomto prípade $\delta$ aby pre každé $x\in(4-\delta,4+\delta)\backslash\{2\}$ platilo $2-\varepsilon <  f(x) < 2+\varepsilon$ t.j. $2-0.01 <  f(x) < 2+0.01$ t.j. $1.99 < f(x) < 2.01$ ? <font color=orange> Nájdite vhodné $\delta$ pre $\varepsilon=0.01$ rovnakým prístupom ako v prípade $\varepsilon=0.5$. Taktiež to overte pomocou interaktívneho kódu. </font>

Uvažujme teraz vo všeobecnosti nejaké $\varepsilon>0$. Predpokladajme, že $f(x)=\sqrt{x}$ je v $\varepsilon$-páse tzn. $|f(x)-2|<\varepsilon$ resp. ekvivalentne $2-\varepsilon<f(x)<2+\varepsilon$. Upravujme ďalej: 

\begin{align*}
|f(x)-2|<&\varepsilon \\ 
-\varepsilon<f(x)-2<&\varepsilon\\
-\varepsilon<\sqrt{x}-2<&\varepsilon \\ 
2-\varepsilon<\sqrt{x}<&2+\varepsilon \\ 
(2-\varepsilon)^2<x<&(2+\varepsilon)^2 \\ 
4-4\varepsilon+\varepsilon^2<x<&4+4\varepsilon+\varepsilon^2 \\
4-(4\varepsilon-\varepsilon^2)<x<&4+(4\varepsilon+\varepsilon^2).
\end{align*}

Posledný krok (nerovnosti), to je len úprava do vhodného tvaru v zmysle "$4-\text{ niečo}<x<4+\text{ niečo}$". Podobne ako v predošlých prípadoch s konkrétnymi hodnotami pre $\varepsilon$, nám teraz posledné nerovnosti ponúkajú dve možnosti pre hľadané $\delta$ t.j. $\delta<4\varepsilon-\varepsilon^2$ alebo $\delta<4\varepsilon+\varepsilon^2$, pričom z nich opäť vyberieme minimum, t.j. $\delta<\min\{4\varepsilon-\varepsilon^2,4\varepsilon+\varepsilon^2\}$ aby boli splnené obe požadované nerovnosti $4-(4\varepsilon-\varepsilon^2)<x<4+(4\varepsilon+\varepsilon^2)$. Keďže $\varepsilon>0$ tak minimum je $4\varepsilon-\varepsilon^2$ a teda potrebujeme zvoliť $\delta<4\varepsilon-\varepsilon^2$.

Keď do tohto všeobecného vyjadrenia $\delta<4\varepsilon-\varepsilon^2$ dosadíme $\varepsilon = 0.5$ alebo $\varepsilon = 0.01$ tak dostaneme tú istú nerovnosť pre hľadané $\delta$ k príslušnému $\varepsilon$ ako v konkrétnych príkladoch vyššie. <font color=orange> Overte to! </font>

Práve sme dokázali nasledovné: <br/>
Ak zvolíme ľubovoľné $\varepsilon>0$ a k nemu vezmeme $\delta<4\varepsilon-\varepsilon^2$, tak potom pre $|x-4|<\delta, x\neq4$ bude platiť $|f(x)-2|<\varepsilon$, čo znamená, že je spĺnená definícia limity funkcie $f(x)=\sqrt{x}$ v bode $x_0=4$ tzn. platí  

$$\lim\limits_{x \to 4} \sqrt{x} = 2.$$

<ins> Riešenie: <ins/>

Využijeme známe fakty z goniometrie (bez zdôvodnenia), konkrétne pre ľubovoľný uhol $\theta$ platí (viď obrázky): 

$$\sin\theta\le\theta\le\tan\theta.$$

 
$$\text{Zdroj: www.mathwarehouse.com.}$$

Upravujme nerovnosti $\sin\theta\le\theta\le\tan\theta$ ďalej pomocou ekvivalentných úprav, s cieľom ohraničiť funkciu $h(\theta)=\dfrac{\sin\theta}{\theta}$ zdola nejakou funkciou $f(\theta)$, zhora nejakou funkciou $g(\theta)$, pričom potrebujeme vedieť ukázať, že $\lim\limits_{\theta \to 0} f(\theta)=\lim\limits_{\theta \to 0} g(\theta)$. 

\begin{align*} 
\sin\theta & \le\theta\le\tan\theta \\ 
\sin\theta & \le\theta\le\dfrac{\sin\theta}{\cos\theta} \\ 
\dfrac{\sin\theta}{\sin\theta} & \le\dfrac{\theta}{\sin\theta}\le\dfrac{\sin\theta}{\cos\theta}\cdot\dfrac{1}{\sin\theta} \\
1 & \le \dfrac{\theta}{\sin\theta}\le\dfrac{1}{\cos\theta} \\
1 & \ge \dfrac{\sin\theta}{\theta}\ge\cos\theta
\end{align*}

Na nejakom prstencovom okolí bodu $0$ vieme funkciu $h(\theta)$ vtesnať resp. "stlačiť" medzi funkcie $f(\theta)$ a $g(\theta)$, ináč povedané, vieme funkciu $h(\theta)$ ohraničiť funkciami $f(\theta)$ a $g(\theta)$ (viď graf nižšie). <br/>
Nemalo by byť ťažké vidieť, že $\lim\limits_{\theta\to0}1=\lim\limits_{\theta\to0}\cos\theta=1$. Podľa Vety 7 teda platí aj 
$\lim\limits_{\theta\to0} \dfrac{\sin\theta}{\theta}=1$.

### Kedy funkcia nemá vlastnú limitu vo vlastnom bode?

Je niekoľko situácii, kedy funkcia nemá limitu v nejakom bode $x_0$, tzn. keď $x \rightarrow x_0$ tak $f(x)$ sa neblíži k jednému konkrétnemu číslu. Môžu to byť napr. tieto tri časté prípady: 

1. Hodnoty funkcie $f(x)$ sa môžu blížiť k dvom rôznym číslam, v závislosti od toho či $x \rightarrow x_0$ sprava alebo zľava (o tomto pojednáva Veta 1); 

2. Funkcia môže rásť nad všetky medze ($f(x)$ ide do $+\infty$) alebo klesať pod všetky možné hodnoty ($f(x)$ ide do $-\infty$), keď $x \rightarrow x_0$;  

3. Funkcia môže oscilovať bez toho, aby sa jej hodnoty blížili k jednému konkrétnemu číslu, keď $x \rightarrow x_0$ (viď Príklad 2 časť b) alebo funkcia v Príklade 12 ). 

Každú z uvedených možností budeme po poradí ilustrovať na príklade.

### <font color=red> Definícia 3 </font>

Nech funkcia $f$ je definovaná na nejakom prstencovom okolí bodu $x_0$ (tzn. nemusí byť definovaná v $x_0$).  Hovoríme, že:

funkcia $f$ má v bode $x_0$ **_nevlastnú limitu_** $+\infty$, ak
$$(\forall c \in \mathbb{R})(\exists \delta > 0)(\forall x \in (x_0 - \delta,x_0+\delta))\ f(x)> c;$$

funkcia $f$ má v bode $x_0$ **_nevlastnú limitu_** $-\infty$, ak
$$(\forall c \in \mathbb{R})(\exists \delta > 0)(\forall x \in (x_0 - \delta,x_0+\delta))\ f(x)< c.$$

---

---

#### Poznámka 5

* Definícia nevlastnej limity $+\infty$ v bode $x_0$ nám v podstate hovorí, že ak zvolíme hodnotu $c$ ľubovoľne veľkú, vždy k nej budeme vedieť nájsť také prstencové okolie bodu $x_0$, že pre $x$ z tohto okolia bude platiť $f(x)>c$. Inými slovami **ako sa blíži $x$ k $x_0$ tak $f(x)$ prevyšuje ľubovoľne veľké číslo**. **Podobne v prípade nevlastnej limity $-\infty$, $f(x)$ klesá pod akúkoľvek hranicu keď $x\to x_0$.**  


* Nevlastné limity sprava a zľava sa  definujú analogicky, ako v prípade vlastnej limity vo vlastnom bode (uvažuje sa $x \in (x_0,x_0+\delta)$ resp. $x \in (x_0 - \delta,x_0)$).

---

**Geometrický význam nevlastnej limity v bode**: pre každú priamku $y = c$  existuje okolo bodu $x_0$ interval polomeru $\delta$ taký, že úsek grafu funkcie vymedzený krajnými bodmi tohto intervalu leží celý nad (resp. pod - v prípade nevlastnej limity $-\infty$) danou priamkou.


$$\text{Zdroj: Madaras (2019).}$$

### <font color=red> Definícia 4 </font>

Nech funkcia $f$ je definovaná na nejakom intervale $(a,\infty)$. Hovoríme, že funkcia $f$ má v bode $+\infty$ limitu rovnú $A$, ak

$$(\forall \varepsilon > 0)(\exists k \in \mathbb{R}^+)(\forall x > k)\ |f(x) - A| < \varepsilon .$$

Ak funkcia $f$ je definovaná na nejakom intervale $(-\infty,a)$, tak hovoríme, že funkcia $f$ má v bode $-\infty$  limitu rovnú $A$, ak

$$(\forall \varepsilon > 0)(\exists k \in \mathbb{R}^-)(\forall x < k)\ |f(x) - A| < \varepsilon .$$

---

---

#### Poznámka 8

* Symbolom $\mathbb{R}^-$ označujeme záporné reálne čísla, podobne symbolom $\mathbb{R}^+$ označujeme kladné reálne čísla. 


* Definícia **limity v nevlastnom bode** (t.j. v bode $+\infty$ alebo $-\infty$) v princípe hovorí, že ak zvolíme ľubovoľné $\varepsilon>0$, vždy k nemu existuje také $k>0$, že pre všetky $x$, ktoré sú (na reálnej osi) napravo od $k$ platí $|f(x)-A|<\varepsilon$, ináč povedané, keď $x$ "ide do nekonečna", tak funkcia $f(x)$ sa blíži k hodnote $A$ (ľubovoľne blízko sa k nej môže priblížiť, podľa toho aké $\varepsilon$ zvolíme). Analogicky je to v prípade, keď $x\to-\infty$, s tým rozdielom, že pozeráme na $x$ naľavo od $k$.  

--- 

**Geometrický význam limity v nevlastnom bode:** pre každú šírku pásu (rovnú $2\varepsilon$, kde $\varepsilon$ je vzdialenosť okraja pásu od jeho čiarkovanej osi) s pozdĺžnou osou $y = A$ (horizontálna prerušovaná čiara) existuje bod $k$ taký, že úsek grafu funkcie napravo (resp. naľavo - v prípade limity pre $x \to -\infty$) leží celý vnútri daného pásu.

 
$$\text{Zdroj: Madaras (2019).}$$

Podľa Definície 5 potrebujeme ukázať, že nech vezmeme akékoľvek reálne číslo $c$, existuje k nemu také kladné číslo $k$, že pre všetky $x$ väčšie ako $k$ bude platiť $f(x)>c$. Postup riešenia rozdelíme na dve časti, najprv sa pozrieme na prípad $c\le0$ a potom na $c>0$. <br/>
Ak by sme uvažovali $c \le 0$, tak keďže $x^2 \ge 0$ pre všetky $x\in\mathbb{R}$ a $x^2=0$ iba pre $x=0$, mohli by sme vziať ľubovoľné $k>0$, napr. $k=1$ a potom pre všetky $x>1$ by platilo $x^2>c$. <br/> 
Uvažujme teraz $c>0$. Položme $k=\sqrt{c+1}$. Potom pre $x>k$ platí nasledovné: $x^2 > k^2 = c+1 > c$ odkiaľ máme $x^2 > c$, čo bolo potrebné dokázať. 


Uvedený postup je možné "obísť" tak, že aplikujeme predposledný poznatok zo zoznamu limít na konci učebného textu. Pri tom je potrebné si uvedomiť, že platí $f(x)=\dfrac{x^2}{1}$, a teda v čitateli zlomku je polynóm druhého stupňa $n=2$, v menovateli je konštanta - t.j. polynóm nultého stupňa, preto $m=0$. 

```python
# limita funkcie x^2 pre x iduce do + nekonecna 
show(limit(x^2,x=infinity))
```

 # <font color=brown> Dôležité limity</font>

V závere ešte sumarizujeme niektoré užitočné **tvrdenia pre výpočet špeciálnych limít**. **Tieto tvrdenia je možné aplikovať pri práci s nevlastnými limitami funkcií vo vlastnom aj nevlastnom bode.** 

--- 

### <font color=green> Veta 11 </font>  

Nech $\lim\limits_{x \to x_0} f(x) = +\infty$, $\lim\limits_{x \to x_0} g(x) = +\infty$ a funkcia $h$ je zdola ohraničená na nejakom prstencovom okolí bodu $x_0$, pričom $x_0\in\mathbb{R}\cup\{+\infty,-\infty\}$. Potom platí

a) $\lim\limits_{x \to x_0} \left(f(x)+g(x)\right) = +\infty$;


b) $\lim\limits_{x \to x_0} \left(f(x)\cdot g(x)\right) = +\infty$; 


c) $\lim\limits_{x \to x_0} \left(f(x)+h(x)\right) = +\infty$;


d) Ak funkcia $h$ je zdola ohraničená kladnou konštantou na nejakom prstencovom okolí bodu $x_0$, tak $\lim\limits_{x \to x_0} \left(f(x)\cdot h(x)\right) = +\infty$. 

---

---

#### Poznámka 11 

Taktiež platia aj alternatívne tvrdenia k tvrdeniam c), d) vo Vete 11: 

* Ak funkcia $h$ je zhora ohraničená na nejakom prstencovom okolí bodu $x_0$ a $\lim\limits_{x \to x_0} f(x) = -\infty$, tak $\lim\limits_{x \to x_0} \left(f(x)+h(x)\right) = -\infty$. 


* Ak funkcia $h$ je zdola ohraničená kladnou konštantou na nejakom prstencovom okolí bodu $x_0$ a $\lim\limits_{x \to x_0} f(x) = -\infty$, tak $\lim\limits_{x \to x_0} \left(f(x)\cdot h(x)\right) = -\infty$.

### <font color=green> Veta 12 </font>  

Ak $\lim\limits_{x \to x_0} |f(x)|=+\infty$, kde $x_0\in\mathbb{R}\cup\{+\infty, -\infty\}$, potom 

$$\lim\limits_{x \to x_0} \frac{1}{f(x)}=0.$$
