# <font color=brown> Motivácia </font>

O limitách funkcií už čo-to vieme, a preto sa v rámci témy 2 zameriame na ďalší dôležitý koncept vychádzajúci práve z limity funkcie. Konrétne pôjde o **spojitosť funkcie**. 

 

**Spojité funkcie** majú dôležitú úlohu napr. **pri modelovaní deformácii pomocou CAD** (Computer-aided design) systémov, taktiež **pri modelovaní priestorových dát** (povrchov) **s využitím GIS** (Geografických Informačných Systémov). Okrem iného so **spojitými funkciami** je možné sa stretnúť napr. aj **pri hlbokom učení** (deep learning) pomocou neuronových sietí, kde sa vyskytujú tzv. **aktivačné funkcie** (napr. sigmoidálna funkcia, tangenciálna funkcia, ReLu ...). Ich úlohou je určitým spôsobom regulovať (normovať) výstupy z neurónov, tzn. aby bol výstup napr. hodnota medzi $0$ a $1$ (niečo ako pravdepodobnosť) ale taktiež určujú do istej miery to, ktorý neurón je ako veľmi dôležitý pre danú predikciu resp. či sa má alebo nemá v sieti aktivovať. Skrátene by sme mohli povedať, že **aktivačné funkcie napomáhajú "učeniu sa" danej neurónovej siete**, aby bola čo najlepšie schopná rozoznávať a klasifikovať určité objekty, robiť predikcie a pod.  

 

V téme 2 teda zavedieme pomocou limity pojem **spojitosti funkcie** v bode aj na intervale. Objasníme základné vlastnosti spojitosti funkcie a všetko budeme ilustrovať na konkrétnych príkladoch. Priblížime si aj **interpretáciu** (tiež geometrickú) **spojitosti** v jednoduchom kontexte. Osobitne sa pozrieme na vlasnosti **spojitej funkcie na uzavretom intervale** (**Weierstrassova veta o ohraničenosti, Weierstrassova veta o maxime/minime, Bolzanova veta**). S využitím **spojitosti** si vysvetlíme **metódu poltenia intervalov** pre numerické riešenie (s ľubovoľnou presnosťou) rovníc, ktorých exaktné korene môže byť obtiažne nájsť.
