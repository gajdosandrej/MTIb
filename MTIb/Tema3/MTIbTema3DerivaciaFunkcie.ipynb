 # <font color=brown> Derivácia funkcie (v bode)</font>

Pri zavádzaní derivácie funkcie v bode, ktorá vyjadruje **ako rýchlo** sa funkcia mení (rastie/klesá) resp. blíži niekam, začneme podobným príkladom o priemernej a okamžitej rýchlosti ako v téme 1 pri limite funkcie, ktorá vyjadrovala **kam** (ku akej hodnote) sa funkcia blíži. 

---

#### <font color=blue> Príklad 1 </font>  

Bežná jazda v zábavnom parku zdvihne jazdcov do určitej výšky, aby im následne umožnila uraziť určitú vzdialenosť voľným pádom pred ich bezpečným zastavením. Predpokladajme, že počas takejto jazdy klesajú jazdci voľným pádom z výšky 150 stôp. Možno si spomeniete z hodín fyziky, že výška (v stopách) jazdca $t$ sekúnd po páde (kvôli jednoduchosti ignorujeme odpor vzduchu a pod.), môže byť celkom dobre modelovaná pomocou funkcie $f(t) = -16t^2 + 150$. 

Keď položíme $f(t)=0$ (nulová výška), tak ľahko zistíme, že bez akéhokoľvek zásahu sa jazdec ocitne na zemi (dopadne na zem) v čase $t=2.5\sqrt{1.5}\approx 3.06$ sekundy. Predpokladajme, že dizajnéri jazdnej dráhy v zábavnom parku sa rozhodnú začať spomaľovať pád jazdcov po 2 sekundách (čo zodpovedá výške 86 stôp). Akú rýchlosť bude mať jazdec práve v tom čase $t=2s$? 

K dispozícii máme funkciu polohy (resp. prejdenej dráhy), ale to, čo chceme vypočítať, je rýchlosť v určitom časovom okamihu, t.j. chceme okamžitú rýchlosť. Predpokladajme, že zatiaľ nevieme ako ju vypočítať (aj keď sme už niečo naznačili v predošlej téme). 

Z bežnej skúsenosti však vieme, ako vypočítať priemernú rýchlosť  - t.j. celková prejdená dráha deleno celkový čas. Napr. ak za 2 hodiny prejdeme 60 km, vieme, že sme mali priemernú rýchlosť 30 km/h lebo 30=60/2.) Toto sme zaviedli už v téme 1 v príklade s padajúcou guľou. Vo všeobecnosti teda, keď potrebujeme vypočítať priemernú rýchlosť medzi časovými okamihmi $t_0$ a $t_1$ t.j. na intervale $\langle t_0,t_1 \rangle$, tak platí 

$$\text{priemerná rýchlosť } = \dfrac{\text{celková dráha}}{\text{celkový čas}} = \dfrac{f(t_1)-f(t_0)}{t_1-t_0} = \dfrac{f(t_0+h)-f(t_0)}{h},$$

pričom poslednú rovnosť dostaneme tak, že $t_1$ nahradíme hodnotou $t_0+h$, $h>0$, tzn. položíme $t_1=t_0+h$, t.j. miesto intervalu $\langle t_0,t_1 \rangle$ môžeme uvažovať interval $\langle t_0,t_0+h \rangle$, je to iba vec označenia. 

Ak $y=f(t)$ tak rozdiel $\Delta t=t_1-t_0$ sa zvykne nazývať **prírastok argumentu** funkcie $f$ a rozdiel $\Delta y=f(t_1)-f(t_0)$ sa zvykne nazývať **prírastok funkcie** funkcie $f$. Potom **relatívny prírastok funkcie** $f$ je podiel 

$$\dfrac{\Delta y}{\Delta t}=\dfrac{f(t_1)-f(t_0)}{t_1-t_0} = \dfrac{f(t_0+h)-f(t_0)}{h}. $$

Z hľadiska prírodných vied relatívny prírastok zodpovedá zmene závislej veličiny $y$ vzhľadom ku zmene nezávislej veličiny $t$. 

Vráťme sa k našej konkrétnej situácii. Môžeme sa pokúsiť aproximovať okamžitú rýchlosť v čase $t_0=2$ pomocou priemernej rýchlosti tak, že vypočítame priemernú rýchlosť na časovom intervale obsahujúcom $t_0=2$. Ak vezmeme dostatočne malý interval okolo hodnoty $t_0=2$, môžeme dostať rozumnú/presnú aproximáciu hľadanej okamžitej rýchlosti v čase $t_0=2$. Táto skutočnosť sa bežne používa. Napríklad vysokorýchlostné kamery sa používajú na sledovanie rýchlo sa pohybujúcich objektov. Vzdialenosti sa merajú na pevnom počte snímok, aby sa vygenerovala presná aproximácia rýchlosti. 

Uvažujme najprv interval od $t_0=2$ po $t_1=3$ (tesne pred tým, ako sa jazdec dotkne zeme resp. ako dopadne na zem). Priemerná rýchlosť na tomto intervale je 

$$\dfrac{f(3)-f(2)}{3-2}=-80 \text{ ft/s (stôp za sekundu)}, $$

pričom znamienko mínus znamená, že jazdec sa pohybuje nadol. Skracovaním intervalu t.j. približovaniu sa k hodnote $t_0=2$ a výpočtom priemernej rýchlosti na kratšom intervale pravdepodobne dostaneme presnejší odhad pre okamžitú rýchlosť v čase $t_0=2$. Na intervale $\langle 2, 2.5 \rangle$ 

$$\dfrac{f(2.5)-f(2)}{2.5-2}=-72 \text{ ft/s}.$$ 

Pre interval $\langle 2, 2.1 \rangle$ máme 

$$\dfrac{f(2.1)-f(2)}{2.1-2}=-65.6 \text{ ft/s}.$$ 

V prípade intervalu $\langle 2, 2.01 \rangle$ je to 

$$\dfrac{f(2.01)-f(2)}{2.01-2}=-64.16 \text{ ft/s}.$$ 

Vo všeobecnosti teda počítame priemernú rýchlosť na intervale $\langle 2, 2+h \rangle$ pre stále menšie a menšie hodnoty $h>0$, t.j. počítame 

$$\dfrac{f(2+h)-f(2)}{h},$$

kde $h>0$ je malé číslo. 

Ideálne by sme potrebovali vziať $h=0$ a vypočítať hodnotu zlomku uvedeného vyššie, avšak to by viedlo k výrazu typu "$0/0$", čo je tzv. neurčitý výraz. Môžeme ale použiť limitu, podobne ako v téme 1 v príklade s rýchlosťou padajúcej gule. 

Na základe aproximácii okamžitej rýchlosti pomocou priemerných rýchostí uvedených vyššie sa zdá, že sa okamžitá rýchlosť v čase $t_0=2$ blíži k hodnote $-64$ ft/s. Priamy výpočet limity vyzerá nasledovne 

$$\lim\limits_{h\to0}\dfrac{f(2+h)-f(2)}{h}=\lim\limits_{h\to0}\dfrac{-16(2+h)^2+150-(-16(2)^2+150)}{h}=\lim\limits_{h\to0}\dfrac{-64h-16h^2}{h}=\lim\limits_{h\to0}\dfrac{h(-64-16h)}{h}=\lim\limits_{h\to0} (-64-16h)=-64.$$

Poslednú limitu vypočítame jednoducho na základe Vety 3 z témy 1 resp. keďže polynomická funkcia $-64-16h$ je spojitá tak limita z nej v bode $0$ je rovná funkčnej hodnote v bode $0$. 

Graficky si môžeme hodnoty priemerných rýchlostí predstaviť ako sklony resp. smernice príslušných priamok (**_sečníc_**), ktoré prechádzajú bodmi so súradnicami $[2,f(2)]$, $[2+h,f(2+h)]$. Viď interaktívna vizualizácia nižšie ...  

Na obrázkoch a), b), c) máme postupne priblížený pohľad na sečnicu (červená priamka) keď $h=1$. Je to kvôli tomu aby sme si všimli, ako zobrazená sečnica (červenou farbou) aproximuje (nadobúda hodnoty blízke $f$) funkciu $f$ (modrou farbou) medzi bodmi $[2,f(2)]$, $[2+h,f(2+h)]$ resp. teraz konkrétne pre $h=1$ t.j. medzi bodmi $[2,f(2)]$, $[3,f(3)]$. Mimochodom sú to dva body, v ktorých sečnica pretína graf funkcie (preto aj ten názov - sečnica). Aproximácia funkcie $f$ v okolí bodu $x_0=2$ pomocou sečnice nie je zlá, no všimnime si, že aproximácia pomocou **_dotyčnice_** v bode $x_0=2$ na obrázku d) je presnejšia/lepšia (dotyčnica viac splýva s funkciou $f$). Dotyčnica je priamka, ktorá sa dotýka grafu funkcie $f$ práve v jednom bode, tu je to konkrétne bod $[2,f(2)]$. 



Nasledujúci interaktívny kód môžete využiť na experimentovanie. Skúste zmenšovať $h$ postupne k nule a všímajte si, ako sa postupne sečnica v bodoch $[2,f(2)]$, $[2+h,f(2+h)]$ mení na dotyčnicu v bode $[2,f(2)]$. Sečnica pre h=0 (čo by už mala byť vlastne dotyčnica) sa nevykreslí, kód vypíše chybu. Je to spôsobené nulou v menovateli zlomku, lebo smernica (sklon) $k$ sečnice je daná vzťahom $k=\dfrac{f(2+h)-f(2)}{(2+h)-2}$. Žiadalo by sa tu poznať hodnotu výrazu $k=\dfrac{f(2+h)-f(2)}{(2+h)-2}$ pre $h=0$, aby sme vypočítali smernicu dotyčnice v bode $[2,f(2)]$ ku grafu funkcie $f$ popisujúcej pád jazdca. Riešením je opäť limita t.j. $\lim\limits_{h\to0}\dfrac{f(2+h)-f(2)}{(2+h)-2}=\lim\limits_{h\to0}\dfrac{f(2+h)-f(2)}{h}$. Takú limitu voláme **_derivácia funkcie_ $f$ v bode $t_0=2$** a je vlastne **rovná smernici $k$ dotyčnice v bode** $[2,f(2)]$ ku grafu funkcie $f$. **Vyjadruje okamžitú zmenu funkcie $f$ v danom bode** (tu konkrétne okamžitú rýchlosť t.j. okamžitú zmenu dráhy podľa času resp. vzhľadom na zmenu času). Naproti tomu **smernica sečnice vyjadruje priemernú zmenu hodnoty funkcie medzi dvoma bodmi** (v ktorých pretína sečnica graf funkcie), tu konkrétne priemernú rýchlosť napr. medzi časovými okamihmi $t_0=2$ a $t_1=3$. 

```python
var('t')
pretty_print(html('<h2>Sečnica grafu funkcie f</h2>')) 
@interact 
def tangent_line(f = input_box(default=-16*t^2+150), tstart = slider(0,10,1/10,0), tkoniec = slider(0,10,1/10,4), t0 = input_box(default=2), h = slider(0, 1, 1/10, 1)): 
        
    # var('t')    
    t1 = t0+h
    y0 = f(t0)
    y1 = f(t1)
    smernica = (y1-y0)/(t1-t0) 
    
    p = plot(f(t), t, tstart, tkoniec) 
    pt1 = point((t0, y0), rgbcolor=(1,0,0), pointsize=30) 
    pt2 = point((t1, y1), rgbcolor=(1,0,0), pointsize=30) 
    
    d = 3
    l = line([(t0-d, y0-d*smernica), (t1+d, y1+d*smernica)], rgbcolor=(1,0,0)) 
    l1 = line([(t0,y0),(t0,0)], color = 'red', linestyle = 'dashed')
    l2 = line([(t0,y0),(0,y0)], color = 'red', linestyle = 'dashed')
    l3 = line([(t1,y1),(t1,0)], color = 'red', linestyle = 'dashed')
    l4 = line([(t1,y1),(0,y1)], color = 'red', linestyle = 'dashed') 
    
    txt1 = text("[%s, %s]" % (float(t0), float(y0)), (t0+0.5, y0-0.8), rgbcolor=(1,0,0)) 
    txt2 = text("[%s, %s]" % (float(t1), float(y1)), (t1+0.5, y1-0.8), rgbcolor=(1,0,0)) 
    txt3 = text("smernica (sklon) $k$ = %s" % float(smernica), ((t0+t1)/2, max(f(tstart),f(tkoniec))+2*d), rgbcolor=(1,0,0)) 
    txt4 = text('$t_0$', (t0,-40), color = 'red')
    txt5 = text('$f(t_0)$', (-0.6,y0), color = 'red') 
    txt6 = text('$t_1=t_0+h$', (t1,-60), color = 'red')
    txt7 = text('$f(t_1)=f(t_0+h)$', (-0.8,y1), color = 'red')
    
    g = p+pt1+pt2+l+txt1+txt2+txt3+l1+l2+l3+l4+txt4+txt5+txt6+txt7
    show(g) 
```

### <font color=red> Definícia 1 </font> 

Nech $f$ je funkcia definovaná na množine $M$ a nech $x_0 \in M$. Hovoríme, že: 

funkcia $f$ má v bode $x_0$ **_deriváciu_** rovnú $f'(x_0)$, ak existuje (vlastná) limita 

$$f'(x_0) =\lim\limits_{x \to x_0} \dfrac{f(x) - f(x_0)}{x - x_0};$$

funkcia $f$ má v bode $x_0$ **_deriváciu sprava_** rovnú $f'_+(x_0)$, ak existuje (vlastná) limita 

$$f'_+(x_0) = \lim\limits_{x \to x_0+} \dfrac{f(x) - f(x_0)}{x - x_0};$$

funkcia $f$ má v bode $x_0$ **_deriváciu zľava_** rovnú $f'_-(x_0)$, ak existuje (vlastná) limita 

$$f'_-(x_0) = \lim\limits_{x \to x_0-} \dfrac{f(x) - f(x_0)}{x - x_0}.$$

#### <font color=blue> Príklad 2 </font> 

Vypočítajte deriváciu funkcie $f(x)=\dfrac{1}{x}$ v bode $x=2$. 

<ins> Riešenie: </ins> 

Podľa Poznámky 2 (vyplývajúcej z definície) môžeme písať 

$$f'(2)=\lim\limits_{h\to0}\dfrac{f(2+h)-f(2)}{h}.$$ 

Dosadíme a upravíme 

$$f'(2)=\lim\limits_{h\to0}\dfrac{\frac{1}{2+h}-\frac{1}{2}}{h}=\lim\limits_{h\to0}\dfrac{1}{h}\left(\dfrac{1}{2+h}-\dfrac{1}{2}\right)=\lim\limits_{h\to0}\dfrac{2-(2+h)}{2h(2+h)}=\lim\limits_{h\to0}\dfrac{-h}{2h(2+h)}.$$ 

Keďže limita súvisí s prstencovým okolím bodu $2$ a nie priamo s bodom $2$, tak $h$ v zlomku môžeme vykrátiť, čím dostávame 

$$f'(2)=\lim\limits_{h\to0}\dfrac{-1}{2(2+h)}=-\dfrac{1}{4}.$$ 

Posledná rovnosť vznikla dosadením $h=0$, keďže $\dfrac{-1}{2(2+h)}$ je už spojitá funkcia premennej $h$ a môžeme sa tak odvolať na Poznámku 1 v téme 2. Alebo si uvedomíme, že čitateľ i menovateľ zlomku $\dfrac{-1}{2(2+h)}$ sú polynomické funkcie premennej $h$, pričom menovateľ je rôzny od $0$ pre $h=0$, potom sa stačí odvolať na Vetu 3 v téme 1.

#### <font color=blue> Príklad 3 </font> 

Pozrieme sa, ako vyzerá derivácia funkcie $f:\ y=|x|$ v rôznych bodoch jej definičného oboru. Pripomenutie: [absolútna hodnota 1](https://www.youtube.com/watch?v=91aGy0Iu9D0),  [absolútna hodnota 2](https://www.youtube.com/watch?v=zc-mvNc9zok). 

Zrejme $\mathcal{D}(f)=\mathbb{R}$. 

Ak $x_0>0$, potom $f'(x_0)=\lim\limits_{x\to x_0}\dfrac{|x|-|x_0|}{x-x_0}=\lim\limits_{x\to x_0}\dfrac{x-x_0}{x-x_0}=\lim\limits_{x\to x_0}1=1$. 

Ak $x_0<0$, potom $f'(x_0)=\lim\limits_{x\to x_0}\dfrac{|x|-|x_0|}{x-x_0}=\lim\limits_{x\to x_0}\dfrac{-x+x_0}{x-x_0}=\lim\limits_{x\to x_0}-\dfrac{x-x_0}{x-x_0}=\lim\limits_{x\to x_0}-1\cdot1=-1$. 

Ak $x_0=0$, potom $f'(0)=\lim\limits_{x\to0}\dfrac{|x|}{x}$, avšak táto limita (derivácia) neexistuje lebo jednostranné limity (derivácie) sú rôzne. Konkrétne $f'_{+}(0)=\lim\limits_{x\to0^+}\dfrac{|x|}{x}=\lim\limits_{x\to0^+}\dfrac{x}{x}=\lim\limits_{x\to0^+}1=1$ a $f'_{-}(0)=\lim\limits_{x\to0^-}\dfrac{|x|}{x}=\lim\limits_{x\to0^-}\dfrac{-x}{x}=\lim\limits_{x\to0^-}-1=-1$. Preto funkcia $f$ nemá deriváciu v bode $0$. 

Ak by sme použili na výpočet jednostranných derivácii druhý tvar, t.j. ten z Poznámky 5 (obsahujúci $h$), vyzeralo by to nasledovne: $f'_+(0) = \lim\limits_{h \to 0+} \dfrac{|0 + h| - |0|}{h} = \lim\limits_{h \to 0+} \dfrac{h}{h} = 1,\quad f'_-(0) = \lim\limits_{h \to 0-} \dfrac{|0 + h| - |0|}{h} = \lim\limits_{h \to 0+} \dfrac{-h}{h} = -1$. 

Všimnime si: Napriek tomu, že funkcia $f$ je spojitá na celom svojom definičnom obore, nemá deriváciu v každom bode definičného oboru. To znamená, že **Veta 2 sa nedá obrátiť resp. neplatí opačná implikácia**: Ak je funkcia spojitá v bode $x_0$, potom má v bode $x_0$ deriváciu. Platí, že ak je funkcia spojitá v bode $x_0$, nemusí v ňom existovať derivácia, tento príklad je toho dôkazom. 

**O neexistencii derivácie funkcie $f$ v bode $x_0=0$ nám napovedá (aj keď len intuitívne) samotný graf (viď nižšie) tejto funkcie. #### Poznámka 3

* Ak sú limity z Definície 1 resp. limity $\lim\limits_{h \to 0} \frac{f(x_0+h)-f(x_0)}{h}$ resp.
$\lim\limits_{h \to 0+} \frac{f(x_0+h)-f(x_0)}{h}$,$\lim\limits_{h \to 0-} \frac{f(x_0+h)-f(x_0)}{h}$ **nevlastné**, hovoríme, že funkcia $f$ má v bode $x_0$ **_nevlastnú deriváciu_** rovnú $+\infty$ resp. $-\infty$. Na grafe nižšie môžeme vidieť, že ak $f'(x_0)=+\infty$ alebo $f'(x_0)=-\infty$, potom dotyčnica ku grafu funkcie $f$ v bode $[x_0,f(x_0)]$ je kolmá na os $x$, tzn. dotyčnica je vlastne priamka (červenou farbou v Príklade 4) s rovnicou $x=x_0$. Ak v bode $x_0$ existujú jednostranné nevlastné derivácie rôznych znamienok, dotyčnica sprava aj zľava v bode $[x_0,f(x_0)]=[x_0,y_0]$ sú kolmé na os $x$, a teda sú totožné. Priamka $x=x_0$ je v takom prípade (pozri Príklad 5) dotyčnicou ku grafu funkcie $f$ v bode $[x_0,f(x_0)]=[x_0,y_0]$. 


* Z definície derivácie a jej geometrickej interpretácie vyplýva, že dotyčnica $t$ ku krivke, t.j. ku grafu funkcie $f$ v bode $[x_0,y_0]=[x_0,f(x_0)]$ existuje, ak existuje derivácia $f'$ v bode $x_0$ a smernica $k$ uvedenej dotyčnice je rovná hodnote $f'(x_0)$. Potom **rovnica dotyčnice má tvar $y - y_0 = f'(x_0)(x - x_0)$, alebo tiež** 

$$t:\ y = y_0 + f'(x_0)(x - x_0).$$ 

Všimnime si, že rozpísaním a upravením vzťahu pre dotyčnicu, spolu s vhodnou substitúciou vlastne dostávame tzv. smernicový tvar priamky $y=kx+q$, spomínaný v Poznámke 1. Platí totiž 

$$t:\ y=f'(x_0)x + \left(y_0-f'(x_0)x_0\right)=kx+q,$$

kde $k=f'(x_0)$, $q=y_0-f'(x_0)x_0$. 


* Ak $f'(x_0) \neq 0$, tak smernica normály (t.j. priamky, ktorá je kolmá na dotyčnicu a prechádza bodom dotyku) je $-\dfrac{1}{f'(x_0)}$ (vyplýva z toho, že súčin smerníc dvoch kolmých priamok je rovný -1), teda **normála má rovnicu** 

$$n:\ y = y_0 - \dfrac{1}{f'(x_0)}(x - x_0).$$


* Ak $f'(x_0) = 0$, tak normála je rovnobežná s osou $y$ a jej rovnica je $x = x_0$.

Na záver už len vynásobíme $f'(g(x))\cdot g'(x)$, čo nám dá podľa Vety 4 hľadanú deriváciu funkcie $h$, 

$$h'(x)=f'(g(x))\cdot g'(x)=\left(3\left(x\cdot\sin\,x\right)^2\right)\cdot\left(\sin\,x+x\cos\,x\right)=3x^2\sin^3x+3x^3\sin^2x\cos\,x=3x^2\sin^2x\left(\sin\,x+x\cos\,x\right).$$

### <font color=green> Veta 6 (Lagrange - o strednej hodnote) </font> 

Nech funkcia $f$ je definovaná a spojitá  na intervale $\langle a,b\rangle$ a nech existuje $f'(x)$ pre všetky $x \in (a,b)$ (vlastná alebo nevlastná). Potom existuje $c \in (a,b)$ také, že $f'(c) = \dfrac{f(b) - f(a)}{b - a}$. 

--- 

--- 

#### Poznámka 5

Geometrická interpretácia Lagrangeovej vety hovorí, že dotyčnica $t$ ku grafu funkcie $f$ v bode $[c,f(c)]$ je rovnobežná s úsečkou spájajúcou body $[a,f(a)]$ a $[b,f(b)]$:

 
$$\text{Zdroj: Madaras (2019).}$$

Spomeňme si, že smernica priamky prechádzajúcej bodmi  $[a,f(a)]$, $[b,f(b)]$ je daná vzťahom $\dfrac{f(b) - f(a)}{b - a}$, na druhej strane už tiež vieme, že smernica dotyčnice ku grafu funkcie v bode $[c,f(c)]$ je rovná práce derivácii $f'(c)$. Podľa Lagrangeovej vety sa tieto smernice rovnajú, preto odpovedajúce priamky s týmito smernicami musia byť rovnobežné (sú nanajvýš posunuté jedna oproti druhej tak, ako to ilustruje obrázok). Voľne povedané, priamky s rovnakou smernicou "idú" resp. "smerujú" rovnakým smerom.

### <font color=green> Veta 7 (Rolle) </font> 

Nech funkcia $f$ je definovaná a spojitá  na intervale $\langle a,b\rangle$, nech existuje $f'(x)$ pre všetky $x \in (a,b)$ (vlastná alebo nevlastná) a nech $f(a) = f(b)$. Potom existuje $c \in (a,b)$ také, že  $f'(c) = 0$. 

--- 

--- 

#### Poznámka 6 

Geometrická interpretácia Rolleho vety hovorí, že dotyčnica $t$ ku grafu funkcie $f$ v bode $[c,f(c)]$ je rovnobežná s osou $x$:

 
$$\text{Zdroj: Madaras (2019).}$$

Už vieme (Poznámka 3), že rovnica dotyčnice $t$ ku grafu funkcie $f$ v bode $[c,f(c)]$ má rovnicu $t:\ y=f(c)+f'(c)\cdot(x-c)$ a keď podľa Vety 2 (Rolle) máme $f'(c)=0$, po dosadení do rovnice dotyčnice $t$ dostávame $t:\,y=f(c)$, čo je priamka rovnobežná s osou $x$.
