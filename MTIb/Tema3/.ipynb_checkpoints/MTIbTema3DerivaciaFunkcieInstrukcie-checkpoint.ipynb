# <font color=brown> Motivácia </font>

Ďalší dôležitý koncept definovaný pomocou limity funkcie je **derivácia funkcie**. 
 


**Derivácie funkcií** majú svoje miesto okrem iného i **v učiacom algoritme pre neurónovú siet**. Slúžia k tomu, aby sa našlo **minimum tzv. stratovej funkcie** a aby sa následne upravili váhy v celej sieti tak, aby ďalšie predikcie či kalsifikácie boli presnejšie. Stručne povedané, vďaka **deriváciám** a ich významu sa neurónová sieť "prestaví" (skoriguje váhy) a v každom kroku tréningu (učenia) sa zlepšuje/spresňuje. **Derivácie** sú užitočné aj **pri riešení** iných **optimalizačných úloh** odbornej praxe či praktických problémov bežného života. Viac si o tom povieme v ďalšej téme tohto kurzu, kde sa budeme zaoberať práve **aplikáciami derivácii**. 



V téme 3 teda zavedieme pojem **derivácie funkcie** vrátane **derivácie sprava/zľava**, popíšeme ich vlastnosti, ktoré budeme ilustrovať na príkladoch. Priblížime si interpretáciu (aj geometrickú) **derivácie** v jednoduchom kontexte. Okrem **výpočtu derivácie z definície (pomocou limity)** ďalej v učebnom texte uvedieme a zároveň budeme ilustrovať na príkladoch niektoré základné pravidlá pre **derivovanie funkcií**. Ukážeme si tiež **výpočet derivácie zloženej funkcie** a **výpočet derivácií vyšších rádov**. V závere uvedieme dôležité **matematické vety (Rolle a Lagrange)**, priamo súvisiace s deriváciou funkcie v bode, ktoré majú pekné geometrické interpretácie a ich dôsledky sa využívajú napr. aj pri odhadovaní funkčných hodnôt.

# <font color=brown> Učebné ciele</font>

### Základné 


*  Definovať a interpretovať (aj geometricky) deriváciu funkcie v bode (tiež deriváciu zľava i sprava). 


* Sformulovať vztah derivácie a jednostranných derivácii. 


* Definovať deriváciu funkcie na množine a derivácie vyšších rádov. 


* Rozumieť vzťahu derivácie a spojitosti funkcie v bode. 


* Mať prehľad o pravidlách pre derivovanie a tiež o deriváciách elementárnych funkcií. 


* Vysloviť, vedieť interpretovať a ilustrovať na konkrétnom príklade Rolleho vetu. 


* Vysloviť, vedieť interpretovať a ilustrovať na konkrétnom príklade Lagrangeovu vetu. 


* Vysloviť a ilustrovať na konkrétnom príklade Vetu o konštantnej funkcii. 

### Pokročilé 


* Z definície nájsť (výpočtom limity) deriváciu funkcie v danom bode. 


* Určiť deriváciu funkcie s využitím pravidiel a "vzorcov" pre derivovanie elementárnych funkcií (úpravy zderivovaných funkcií aj pomocou Sage). 


* Určiť deriváciu zloženej funkcie i spojitej, rýdzomonotónnej funkcie. 


* Nájsť derivácie vyšších rádov pre zadané funkcie. 


* Aplikovať derivacie na nájdenie dotyčnice a normály v danom bode, so špecifickými vlastnosťami. 


* Overiť, že daná funkcia vyhovuje (je riešením) rovnici, kde vystupuje derivácia funkcie. 


* Nájsť rovnicu dotyčnice a normály ku grafu funkcie s danými vlastnosťami. 


* Nájsť príklady funkcií s určitými vlastnosťami týkajúcimi sa derivácie či spojitosti (napr. funkcia (ne)spĺňajúca predpoklady nejakej vety a pod.).
