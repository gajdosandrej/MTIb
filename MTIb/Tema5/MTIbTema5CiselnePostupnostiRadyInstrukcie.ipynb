<a id=motivacia></a>
# <font color=brown> Motivácia</font>


 

**Algoritmus** je v jednoduchosti chápaný ako **postupnosť krokov**. Od zazvonenia budíka po raňajky zvyčajne človek ráno vykoná tiež určitú postupnosť krokov (viď obrázok). Téma 5 bude ale venovaná špeciálnym - **číselným postupnosťiam** (viď napr. obrázky s bodkami nižšie). Ako **(nekonečné) súčty členov číselných postupností** budeme vytvárať **(nekonečné) číselné rady**. 

  

**Nekonečné číselné rady** sú užitočné okrem iného aj pri **konverzii zlomkov na desatinné čísla** s nekonečným desatinným rozvojom (resp. s nekonečným počtom cifier za desatinnou čiarkou). Tak napr. $1/3 = 0.3333333...$, čo je možné vyjadriť v tvare $3/10 + 3/100 + 3/1000 + 3/10000 + ...$ a ako sa dozvieme, platí  $3/10 + 3/100 + 3/1000 + 3/10000 + ... = 1/3$.<br/> 
Ďalší praktický význam **nekonečných radov** spočíva v tzv. **harmonickej analýze signálov**. Zjednodušene vzaté, signály (zvukové či obrazové) je možné reprezentovať pomocou radov a následne robiť napr. filtrovanie signálov - pri zvukovej nahrávke je možné odstrániť basy alebo ich zosilniť. <br/> 
Nie menej dôležitá je **úloha radov pri aproximácii** (komplikovaných) **funkcií** a **pri určovaní chyby numerickej aproximácie**.  

Funkciami, ich vlastnosťami a tiež limitami sme sa zaoberali na začiatku aktuálneho semestra. Teraz nadviažeme a budeme sa venovať špeciálnemu typu funkcií - **číselným postupnostiam**. Pripomenieme si niektoré ich vlastnosti ako napr. **ohraničenosť**, **monotónnosť** a súvis s **limitami**.  

 

Na základe **číselných postupností** a ich **limít** vybudujeme koncept **nekonečného číselného radu** a jeho **súčtu**. Na príkladoch budeme ilustrovať použitie vybraných **kritérii konvergencie nekonečných radov**, ktoré nám pomôžu rozhodnúť, či daný **rad** má **konečný súčet** alebo nie. <br/> 
Pozrieme sa detailnejšie aj na vybrané konkrétne **špeciálne nekonečné rady** resp. **skupiny radov** (**geometrický rad**, **harmonický rad**, **rady s nezápornými členmi**), s ktorými je možné sa stretnúť pri riešení rôznych úloh.
