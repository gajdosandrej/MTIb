<a id=motivacia></a>
# <font color=brown> Motivácia</font>

**Symetria** je všadeprítomný jav (napr. aj **v Rubikovej kocke**), ako v prírodnom aj tak aj v človekom (vrátane biologického) vytvorenom prostredí. Ľudia majú vrodenú schopnosť vnímať a využívať symetriu v každodennom živote, ale nemusí byť úplne jasné, ako automatizovať toto vnímanie pre človekom vytvorené inteligentné bytosti - napr. roboty. V tradičných vedných disciplínach  zohralo pochopenie symetrie významnú úlohu pri niekoľkých dôležitých objavoch vrátane **teórie relativity** (symetria času a priestoru), pri **štruktúre ľudskej DNA** (dvojitá špirála), v **kvázi kryštáloch** a v ich matematickom náprotivku - tzv. [Penrosove dláždenie (pokrytie)](https://dml.cz/bitstream/handle/10338.dmlcz/141885/PokrokyMFA_54-2009-1_6.pdf). Uvažovanie o symetrii môze zohrávať kľúčovú úlohu v rozvoji umelej resp. strojovej inteligencie. 

 

Výpočtový model symetrie sa týka najmä **robotiky**, **počítačového videnia** a **strojovej inteligencie** vo všeobecnosti, pretože v týchto oblastiach sa študuje, ako môže človekom vytvorená inteligentná bytosť čo najefektívnejšie vnímať a interagovať s (chaotickým) skutočným svetom. **Rozpoznanie symetrií** je prvým krokom k zachyteniu základnej štruktúry problému v reálnom svete a k minimalizácii nadbytočností, ktorý **môže** často **viesť k** podstatným **redukciám vo výpočtoch**. Jedným zo základných obmedzení počítačov je ich konečná presnosť (reprezentácia čísel s určitou konečnou presnosťou). Jedna malá chyba resp. nepresnosť v zaokrúľovaní čísel môže zničiť akúkoľvek dokonalú symetriu. Schopnosť človeka tolerovať odklon od dokonalej symetrie odráža jeho úroveň sofistikovanosti vo vnímaní, ktorú je potrebné zabudovať do rozvoja strojovej/umelej inteligencie. 

 

**Grupy** teda nachádzajú svoje uplatnenie napr. **v robotike, v kryptografii, vo fyzike (Lieove grupy), v chémií, v materiálovom inžinierstve či  v počítačovej grafike (grupy zobrazení/transformácií)** atď. Zaujímavé je tiež napr. využitie teórie grúp **pri návrhu efektívnych algoritmov** (násobenie matíc). 

 

V tejto záverečnej téme zimného semestra si vysvetlíme, ako je možné všeobecnejšie (abstraktnejšie) nazerať na objekty, s ktorými sme pracovali doteraz - ako napr. čísla, zvyškové triedy, matice, funkcie resp. zobrazenia (permutácie/symetrie). Ukážeme si, že uvedené objekty spolu s vhodnými **binárnymi operáciami** môžu za určitých podmienok vytvárať užitočné **matematické štruktúry** - tzv. **grupy** 

 

Pozrieme sa aj na to, ako je možné znázorniť operácie medzi prvkami grupy pomocou **Cayleyho tabuľky**. Poukážeme na špeciálnu vlasnosť týchto tabuliek, ktorá súvisí s tzv. **Latinskými štvorcami** a tak trochu aj so Sudoku. Objasníme, čo to znamená, že dve grupy je možné stotožniť (**izomorfné grupy**). Zavedieme i ďalšie dôležité pojmy súvisiace s grupami ako napr. **podgrupa**, **rád grupy** či **rád prvku grupy**. Zadefinujeme dôležité typy aj vlastnosti grúp a vyústi to až k **Lagrangeovej vete** či **Cayleyho vete**. Samozrejme, ako tomu bolo i doteraz, všetko bude doplnené veľkým mnoštvom (ilustračných) príkladov a ukážeme si tiež základnú prácu s grupami v Sage. Na záver veľmi stručne spomenieme aj o niečo komplikovanejšiu algebraickú štruktúru - tzv. **pole**.
