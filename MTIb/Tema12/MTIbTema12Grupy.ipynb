 # <font color=brown> Úvod </font> 

**Abstraktná algebra** sa zaoberá **algebraickými systémami** (**štruktúrami** - napr. zvyškové triedy pri danom module s operáciou sčítania a/alebo násobenia atď.) v abstraktnejšom resp. všeobecnejšom zmysle, študuje ich vlastnosti, ktoré sa následne častokrát premietajú aj do aplikácií. Okrem zvyškových tried z prvej polovice semestra poznáme rôzne algebraické systémy zodpovedajúce rôznym číselným systémom ako napr. celé čísla $\mathbb{Z}=\{\ldots,-3,-2,-1,0,1,2,3,\ldots\}$, racionálne čísla $\mathbb{Q}=\{\frac{m}{n};\,m,n\in\mathbb{Z}, n\neq0\}$, reálne čísla $\mathbb{R}$ alebo komplexné čísla $\mathbb{C}=\{x+iy; x,y\in\mathbb{R}\}$ (kde $i^2=-1$) a na týchto číselných množinách uvažujeme aj algebraické (aritmetické) operácie ako napr. sčítanie, odčítanie, násobenie. 

Taktiež vieme, že **algebraické operácie** je možné realizovať aj s ďalšími objektmi ako napr. vektory, matice, zvyškové triedy, funkcie (resp. zobrazenia) a to konkrétne sčítavať ich, odčítavať, násobiť prípadne skladať. Niekedy tieto operácie s uvedenými objektmi spĺňajú vlastnosti, na ktoré sme zvyknutí u číselných systémov resp. čísel ale nie vždy - napr. násobenie matíc nie je vo všeobecnosti komutatívne, zatiaľ čo násobenie čísel je komutatívne. Zrejme pre akékoľvek dve reálne čísla $a,b$ platí, že $a \cdot b = b \cdot a$. Na druhej strane ak vezmeme napr. matice $\mathbf{A}=\begin{pmatrix}0&1\\0&0\end{pmatrix}$, $\mathbf{B}=\begin{pmatrix}0&0\\1&0\end{pmatrix}$ tak platí 

$$\mathbf{A}\cdot\mathbf{B}=\begin{pmatrix}0&1\\0&0\end{pmatrix}\cdot\begin{pmatrix}0&0\\1&0\end{pmatrix}=\begin{pmatrix}1&0\\0&0\end{pmatrix}\neq\begin{pmatrix}0&0\\0&1\end{pmatrix}=\begin{pmatrix}0&0\\1&0\end{pmatrix}\cdot\begin{pmatrix}0&1\\0&0\end{pmatrix}=\mathbf{B}\cdot\mathbf{A}.$$ 

**Abstraktná algebra** študuje všeobecné algebraické systémy v axiomatickom rámci, takže dokázané vety platia v čo najširšom možnom rozsahu. Často objavujúce sa algebraické systémy sú _grupy_, _okruhy_, _polia_. My sa trochu oboznámime s grupami a aspoň si povieme, čo je to pole. Pri oboch spomenutých štruktúrach platí, že ich prvkami môžu byť nie len čísla (spolu so štandardnými operáciami) ale prvkami grupy či poľa môžu byť už spomínané vektory, matice, funkcie (resp. zobrazenia) a pod. (spolu s príslušne definovanými operáciami).  

Pred tým, než zadefinujeme exaktne grupu a tiež všetky dôležité pojmy s tým súvsiace, uveďme **niekoľko príkladov grúp** pre lepšiu predstavu. Množina celých čísel $\mathbb{Z}$ vybavená operáciou sčítania je príkladom grupy. Ďalej množiny $\mathbb{Q}$, $\mathbb{R}$, $\mathbb{C}$ spolu s operáciou sčítania čísel sú príkladmi grúp. Množina všetkých dvojrozmerných alebo trojrozmerných (vo všeobecnosti $n$-rozmerných) vektorov (t.j. usporiadaných dvojíc/trojíc alebo $n$-tíc reálnych čísel) vrátane operácie sčítania vektorov (po zložkách) tvorí grupu. <br/> 
Iné dôležité príklady grúp súvisia so symetriami (geometrických) objektov. Tie môžu byť realizované vo všetkých dimenziách ale kvôli názornosti ostaneme v dvoch rozmeroch. Uvažujme rovnoramenný trojuholník ako na obrázku. 

 

Tento trojuholník má jednu os symetrie (súmernosti) - vyznačenú prerušovanou čiarou, ktorá rozdeľuje rovnoramenný trojuholník na dve rovnaké časti. Môžeme si to predstaviť tak, ze ak by sme mali takýto trojuholník narysovaný na papieri a v ňom vyznačenú os, tak presným prehnutím papiera cez túto os symetrie by jednotlivé časti trojuholníka dokonale priliehali k sebe tzn. presne by sa prekrývali. Inými slovami, ak by sme urobili zrkadlový obraz pravej a ľavej časti trojuholníka vzhľadom na vyznačenú os, tak by sme dostali opäť ten istý trojuholník (vrchol $C$ - ten, cez ktorý prechádza os by sa v osovej súmernosti zobrazil sám do seba, ďalej vrchol $A$ by sa zobrazil do vrchola $B$ a vrchol $B$ by sa zobrazil do vrchola $A$ - t.j. zrkadlovo by sa preklopili cez vyznačenú os). Uvedenú osovú súmernosť je teda možné chápať ako zobrazenie a môžeme ho označiť $o$, pričom platí $o(A)=B$, $o(B)=A$, $o(C)=C$, čo môžeme zapísať podobne ako permutácie (pozri tému 5) t.j. $o=\begin{pmatrix}A&B&C\\B&A&C\end{pmatrix}$. Identickým zobrazením (identitou) by bolo také zobrazenie - označme ho $e$, že každý vrchol trojuholníka sa zobrazí sám do seba t.j. $e(A)=A$, $e(B)=B$, $e(C)=C$ resp. $e=\begin{pmatrix}A&B&C\\A&B&C\end{pmatrix}$. Tieto zobrazenia je možné skladať podobne ako permutácie resp. funkcie. Čiže $e\circ o$ by znamenalo, že najprv sa urobí osová súmernosť a následne sa aplikuje identické zobrazenie t.j. $(e\circ o)(A)=e(o(A))=e(B)=B$, $(e\circ o)(B)=e(o(B))=e(A)=A$, $(e\circ o)(C)=e(o(C))=e(C)=C$, čo je vlastne to isté ako zápis $\begin{pmatrix}A&B&C\\B&A&C\\B&A&C\end{pmatrix}$, Vidíme, že aplikovaním identického zobrazenia sa označenie vrcholov trojuholníka nezmení a ostane rovnaké ako po aplikovaní iba samotnej osovej súmernosti, z čoho vyplýva, že $e\circ o=o$. Súhrne to ešte môžeme zapísať takto: 

$$e\circ o=\begin{pmatrix}A&B&C\\B&A&C\\B&A&C\end{pmatrix}=\begin{pmatrix}A&B&C\\B&A&C\end{pmatrix}=o.$$

**Množina $\{e,o\}$ dvoch uvedených zobrazení v rovnoramennom trojuholníku spolu s operáciou $\circ$ skladania zobrazení tvorí grupu.** <br/> 
Niektoré útvary majú ale viacero symetrií - hovoríme, že sú viac symetrické. Napr. rovnostranný trojuholník znázornený na obrázku má tri osi súmernosti resp. symetrie, z ktorých každá ho rozdeľuje na dve idetické časti. 

 

Ak by sme aplikovali dve osové súmernosti po sebe (tzn. zložili dve zobrazenia resp. symetrie podobne ako pred tým pri rovnoramennom trojuholníku), tak by sme vlastne dostali otočenie (buď v smere alebo proti smeru hodinových ručičiek) trojuholníka okolo jeho stredu (resp. ťažiska) o uhol $2\pi/3$ resp. o $120^\circ$ (tomu by zodpovedalo zobrazenie vrcholov v trojuholníku postupne podľa dvoch zvolených osí súmerností). Tieto otočenia (rotácie) rovnostranného trojuholníka sa tiež považujú za zobrazenia resp. symetrie. Nasledujúci obrázok ilustruje výsledok operácie $o_1\circ o_2$. Najprv teda aplikujeme osovú súmernosť podľa osi $o_2$ a hneď na to osovú súmernosť podľa osi $o_1$. Ako sa zobrazia jednotlivé vrcholy trojuholníka si môžeme rozpísať takto: $(o_1\circ o_2)(A)=o_1(o_2(A))=o_1(C)=B$, $(o_1\circ o_2)(B)=o_1(o_2(B))=o_1(B)=C$, $(o_1\circ o_2)(C)=o_1(o_2(C))=o_1(A)=A$ t.j. 

$$o_1\circ o_2=\begin{pmatrix}A&B&C\\C&B&A\\B&C&A\end{pmatrix}=\begin{pmatrix}A&B&C\\B&C&A\end{pmatrix}=r_1.$$

Keď to zhrnieme, tak vlastne dostávame, že $o_1\circ o_2=r_1$, kde $r_1$ je otočenie (rotácia) trojuholníka o $120^\circ$ v smere hodinových ručičiek.   



Každá **symetria** (rovnostranného) trojuholníka sa dá chápať ako určité **zobrazenie** trojuholníka na seba samého (resp. zobrazenie množiny jeho vrcholov do tej istej množiny vrcholov). Celkovo teda existuje šesť symetrií: tri osové súmernosti ($o_1,o_2,o_3$), dve otočenia resp. rotácie ($r_1,r_2$) a identické zobrazenie ($e$). Zloženie dvoch symetrií trojuholníka tzn. ich postupná aplikácia je opäť symetria (jedna zo šiestich uvedených). Táto **kolekcia všetkých šiestich symetrií $\{o_1,o_2,o_3,r_1,r_2,e\}$ spolu s operáciou skladania zobrazení je známa ako _grupa symetrií_ trojuholníka**. Tieto symetrie máme všetky znázornené na nasledujúcom obrázku a sú tiež zapísané podobným štýlom ako sme zapisovali permutácie (viď téma 5).  

 

V princípe tu ide o permutácie trojprvkovej množiny $\{A,B,C\}$. 

Podobne je možné uvažovať i **symetrie obdĺžnika**. Celkovo sú štyri - viď nasledujúci obrázok: identita, otočenie o $180^\circ$, súmernosť podľa vertikálnej osi a súmernosť podľa horizontálnej osi. 

 

**Symetriou geometrického úvaru** (objektu) v podstate rozumieme také zobrazenie (preusporiadanie jeho vrcholov), ktoré zachováva  susednosť vrcholov, príslušnosť strán (k vrcholom), vzdialenosti a uhly. Napr. otočenie (rotácia) obdĺžnika v rovine o $180^\circ$ alebo o $360^\circ$ vráti obdĺžnik v rovine s rovnakou orientáciou ako pôvodný obdĺžnik a tiež s rovnakými vzťahmi medzi vrcholmi. Podobne je to s osovými súmernosťami (vzhľadom na vertikálnu alebo horizontálnu os obdĺžnika). Toto všetko sú teda symetrie obdĺžnika (viď predošlý obrázok). Naproti tomu otočenie (rotácia) odbĺžnik o $90^\circ$ v jednom alebo v druhom smere nemôže byť považované za symetriu, pokiaľ sa nejedná špeciálne o štvorec. Totiž otočeniu o $90^\circ$ v smere hodinových ručičiek by zodpovedalo takéto zobrazenie vrcholov obdĺžika: $A\to B$, $B\to C$, $C\to D$, $D\to A$ (viď nasledujúci obrázok) a tým pádom by sa zmenila napr. dĺžka strany $AB$ (v novom označení vrcholov) oproti pôvodnému obdĺžniku (identita). 

 

Otočenie obdĺžnika o $90^\circ$ si môžeme predstaviť aj tak, že s označením jeho vrcholov nehýbeme a otočíme len samotný 
odĺžnik (viď nasledovný obrázok).

 # <font color=brown> Grupy </font> 

Predchádzajúce príklady grúp ilustrujú, že **každá grupa má dve dôležité charakteristiky**. **Prvou je množina určitých  objektov/prvkov** (napr. čísel, vektorov, funkcií, symetrií, permutácií, ...) a **druhou je metóda, ako sa majú skombinovať dva prvky z danej množiny aby vznikol tretí prvok opäť patriaci do tejto množiny** (napr. pomocou sčítania, násobenia, skladania funkcií resp. zobrazení (symetrií či permutácií) ). Druhá z uvedených charakteristík grupy sa nazýva _binárna operácia_ (na nejakej množine $S$) a exaktne ju zavádza nasledujúca definícia. 

---

### <font color=red> Definícia 1 </font>

Nech $S$ je množina. **_Binárnou operáciou_** $\ast$ na množine $S$ rozumieme zobrazenie $S\times S\rightarrow S:\quad (x,y)\rightarrow x\ast y$, t.j. zobrazenie také, ktoré dvom prvkom $x,y$ z množiny $S$, priradí jeden prvok $x\ast y$ z množiny $S$. 

---

--- 

#### <font color=blue> Príklad 1 </font>

Príklady binárnych operácii $\ast$ na množine $S$: 

1. V Definícii 1 je možné nahradiť $\ast$ napr. aritmetickou operáciou $+$ alebo $-$ alebo $\times$ na vhodnej množine $S$, napr. $S=\mathbb{Q}$ (racionálne čísla) alebo $S=\mathbb{R}$ (reálne čísla). 


2. Miesto $\ast$ v Definícii 1 je možné uvažovať napr. sčítanie alebo násobenie matíc na množine $S$, ktorá bude v tomto prípade predstavovať množinu všetkých matíc typu $n\times n$, kde $n\in\mathbb{N}$. 


3. Za operáciu $\ast$ v Definícii 1 je možné zvoliť napr. sčítanie alebo odčítanie vektorov, pričom množina $S$ bude množina větkých usporiadaných $n$-tíc reálnych čísel ($n$-rozmerných vektorov) t.j. $S=\mathbb{R}^n$. 


4. V Definícii je možné nahradiť operáciu $\ast$ napr. vektorovým súčinom dvoch trojrozmerných vektorov t.j. $(a,b,c)\times(x,y,z)=(bz-cy,cx-az,ay-bx)$ a to je binárna operácia na množine $S=\mathbb{R}^3$ (všetkých usporiadaných trojíc reálnych čísel t.j. trojrozmerných vektorov). 


5. Za operáciu $\ast$ v Definícii 1 je možné zvoliť napr. operáciu skladania zobrazení, konkrétne napr. skladanie symetrií trouholníka alebo štvorca alebo kocky ... pričom množina $S$ potom bude množina (všetkých) symetrií trojuholníka alebo štvorca alebo kocky ... 

---

---

#### Poznámka 1 

* Časť definície binárnej operácie na nejakej množine $S$ hovorí o tom, že výsledkom binárnej operácie na množine $S$ je opäť prvok množiny $S$ tzn. $x\ast y\in S$ pre $x\in S$ aj $y\in S$. Táto vlastnosť sa niekedy popisuje slovami, že **"množina $S$ je uzavretá vzhľadom k operácii $\ast$"**, prípadne alternatívne vyjadrenie **"operácia $\ast$ je uzavretá na množine $S$"**. Tento pojem naberá na význame, keď sa obmedzíme na nejakú podmnožinu $T$ množiny $S$ (kde bola operácia pôvodne definovaná) t.j. $T\subset S$. 


* Ak $T\subset S$ a $\ast$ je binárna operácia na $S$, potom $\ast$ je zobrazenie $S\times S\to S$ a $T\times T$ je podmnožina množiny $S\times S$ (t.j. usporiadané dvojice prvkov z množiny $S$), takže má zmysel uvažovať **zúženie zobrazenia** $\ast: S\times S\to S$ **na množinu** $T\times T$ (všetky usporiadané dvojice prvkov z $T$). Pre prvky $x,y\in T$ určite platí, že $x\ast y\in S$, avšak vo všeobecnosti nemusí pre také prvky platiť, že $x\ast y\in T$. Hovoríme, že podmnožina $T\subset S$ je uzavretá vzhľadom na operáciu $\ast$, ak platí $\left(\forall x,y\in T\right)\quad x\ast y\in T$.  


* Ak je $T\subset S$ uzavretá vzhľadom na $\ast$, potom môžeme uvažovať zúženie operácie $\ast$ na množinu $T\times T$ ako zobrazenie $T\times T\to T$, tzn. ako binárnu operáciu na množine $T$. 


* Niekedy sa používa **skrátený zápis** a binárna operácia sa vynecháva, tzn. $a\ast b$ značí to isté čo $ab$. 

---

--- 

#### <font color=blue> Príklad 2 </font>

1. Množina všetkých párnych celých čísel $2\mathbb{Z}$ **je uzavretá** vzhľadom na binárnu operáciu sčítania. Ináč povedané, súčet dvoch celých párnych čísel je vždy párne celé číslo. Vo všeobecnosti pre súčet dvoch párnych čísel $2m$, $2n$ platí $2m+2n=2(m+n)\in2\mathbb{Z}$, kde $m,n\mathbb{Z}$. 


2. Množina všetkých nepárnych celých čísel $\mathbb{Z}\setminus2\mathbb{Z}$ **nie je uzavretá** vzhľadom na operáciu sčítania. Napr. $5$ a $-13$ sú nepárne celé čísla ale ich súčet $5+(-13)=-8$ je párne celé číslo a teda mimo množiny všetkých nepárnych celých čísel $\mathbb{Z}\setminus2\mathbb{Z}$ (výsledok sčítania dvoch prvkov z $\mathbb{Z}\setminus2\mathbb{Z}$ "vypadol" z množiny $\mathbb{Z}\setminus2\mathbb{Z}$). 

---

Binárna operácia $\ast$ na konečnej množine $S$ sa dá znázorniť vo forme poľa/tabuľky tzv. **_Cayleyho tabuľky_**. Spomeňte si na farebné tabuľky pri modulárnej aritmetike a kongruenciách. 

Ak má množina $S$ $n$ prvkov, potom Cayelho tabuľka ma rozmery $n\times n$, kde každý riadok aj každý stĺpec prislúcha práve jednému prvku z množiny $S$. Prvok v tabuľke na pozícii $[i,j]$ (v $i$-tom riadku a v $j$-tom stĺpci) zodpovedá prvku $x_i\ast x_j$, pričom $x_i,x_j\in S$. 

Pre ilustráciu uvádzame jednoduchý príklad, kde $S=\{0,1\}$ a $\ast$ je obyčajné násobenie čísel. 

 

<font color=orange> Vytvorte Cayleyho tabuľky pre nasledujúce množiny prvkov a binárne operácie: množina celých čísel $\{0,1,2,3\}$ s operáciou sčítana; množina celých čísel $\{0,1,2,3\}$ s operáciou násobenia; množina zvyškových tried $\{0_3,1_3,2_3\}$ s operáciou sčítana modulo $3$; množina zvyškových tried $\{0_3,1_3,2_3\}$ s operáciou násobenia modulo $3$. </font>

Uvažujme teraz pre jednoduchosť $\mathbb{Z}_2=\left\{0_2,1_2\right\}$ spolu s operáciou sčítania $\oplus$. Pozrime sa, aké vlastnosti má systém resp. štruktúra (dvojica) $\left(\mathbb{Z}_2,\oplus\right)$. 

* Uzavretosť operácie $\oplus$, tzn. ak sčítame dva prvky zo $\mathbb{Z}_2$ tak vždy dostaneme len prvok, ktorý patrí do $\mathbb{Z}_2$. Vzhľadom ku komutativite operácie $\oplus$ (nezáleží na poradí prvkov pri sčítaní) máme tieto možnosti: $0_2\oplus0_2=0_2$, $0_2\oplus1_2=1_2$, $1_2\oplus1_2=0_2$. 


* Asociatívnosť operácie $\oplus$, tzn. prvky pri sčítaní môžeme ľubovoľne ozátvorkovať a nemá to vplyv na výsledok sčítania. Platnosť tejto vlastnosti sme videli na príkladoch uvedených vyššie. 


* Pre každý prvok množiny $\mathbb{Z}_2$ existuje tzv. identita (identický prvok), tzn. taký prvok, ktorý keď pripočítame k zvolenému prvku zo $\mathbb{Z}_2$ tak sa ten zvolený prvok nezmení. V prípade $\mathbb{Z}_2$ je identitou (pre všetky prvky zo $\mathbb{Z}_2$) práve prvok $0_2$ lebo $0_2\oplus0_2=0_2$ a $1_2\oplus0_2=1_2$. 


* Pre každý prvok množiny $\mathbb{Z}_2$ existuje tzv. inverzia (inverzný prvok), tzn. taký prvok, ktorý keď pripočítame k zvolenému prvku zo $\mathbb{Z}_2$ tak dostaneme identitu - v tomto prípade dostaneme $0_2$. K prvku $0_2$ je inverzný prvok $0_2$ (je sám k sebe inverzný) lebo $0_2\oplus0_2=0_2$ a k prvku $1_2$ je inverzný prvok $1_2$ (tiež je sám k sebe inverzný) lebo $1_2\oplus1_2=0_2$. 


Tento ale aj predošlé príklady z úvodu nás vedú k nasledujúcej definícii grupy s tým, že teraz miesto $\mathbb{Z}_2$ budeme  uvažovať nejakú všeobecnú monžinu $G$ a miesto operácie $\oplus$ nejakú binárnu operáciu $\ast$ definovanú na množine $G$ (medzi jej prvkami). 

---

### <font color=red> Definícia 2 </font>

Dvojicu $(G,\ast)$, kde $G$ je množina a $\ast$ je binárna operácia na $G$, nazývame **_grupa_**, ak spĺňa nasledujúce tri vlastnosti: 

1. $\ast$ je _asociatívna_, tzn. $$\left(\forall x,y,z\in G\right)\quad x\ast (y\ast z)=(x\ast y) \ast z.$$ 
<br/>
2. $G$ obsahuje prvok $e$ (resp. $e_G$), ktorý je _identitou_ (identickým prvkom) vzhľadom na binárnu operáciu $\ast$, tzn. 
<br/>
<br/>
$$\left(\forall x\in G\right)\quad x\ast e = x = e\ast x.$$
<br/>
3. Každý prvok $x\in G$ má _inverziu_ (inverzný prvok) $\bar{x}$ (resp. $x^{-1}$) v $G$, tzn.
<br/>
<br/>
$$x\ast x^{-1}=e=x^{-1}\ast x$$ 

---

#### <font color=blue> Príklad 3 </font>

#### <font color=blue> Príklad 6 </font>  

Uvažujme Cayleyho tabuľky dvoch grúp ako je to znázornené na nasledujúcom obrázku.

Teda môžeme nazerať na tieto dve grupy ako na totožné - až na pomenovanie (označenie) prvkov. Ak nastáva tento popísaný jav medzi dvoma grupami, hovoríme, že dve grupy sú _izomorfné_ a považujeme ich za totožné (maju rovnakú štruktúru). 

Funkcia, ktorá preznačí prvky jednej grupy na prvky druhej grupy je vlastne bijekcia medzi množinami dvoch grúp. Táto funkcia sa zvykne nazývať _izomorfizmus_. Teraz nasleduje formálna definícia izomorfizmu.

### <font color=red> Definícia 5 </font>

Nech $(G,\ast)$ a $(H,\diamondsuit)$ sú dve grupy, pričom prvá z nich pozostáva z množiny $G$ a z binárnej operácie $\ast$ na tejto množine, zatiaľ čo druhá pozostáva z množiny $H$ a z binárnej operácie $\diamondsuit$ na príslušnej množine. **_Homomorfizmus_** z $(G\ast)$ do $(H,\diamondsuit)$ je také zobrazenie $f:\,G\to H$, pre ktoré platí, že 

$$(\forall x,y\in G)\quad f(x\ast y)=f(x)\diamondsuit f(y).$$

**_Izomorfizmus_** z $(G\ast)$ do $(H,\diamondsuit)$ je bijektívny homomorfizmus $f:\,G\to H$. Ak existuje izomorfizmus z $(G\ast)$ do $(H,\diamondsuit)$, potom hovoríme, že grupy $(G\ast)$, $(H,\diamondsuit)$ sú izomorfné a označujeme to buď $(G\ast) \cong (H,\diamondsuit)$ alebo skrátene $G \cong H$.

Spomenutá cyklická podgrupa sa zostrojí nasledovne: Najprv sa definujú _mocniny_ prvku $a$ v grupe $G$, $a^n\in G$ pre $n\in\mathbb{Z}$. Pre $n<0$ definujeme $a^n=\bar{a}^{-n}$, kde $\bar{a}$ je inverzný prvok k prvku $a$ v grupe $(G,\ast)$.

**_Cyklická podgrupa_** grupy $(G,\ast)$ generovaná prvkom $a$ je podmnožina $\langle a\rangle=\left\{a^n;\,n\in\mathbb{Z}\right\}$ množiny $G$.

#### Poznámka 6

 # <font color=brown> Polia </font> 

O niečo zložitejšia štruktúra ako grupa je tzv _pole_. Pole je tvorené množinou a dvoma binárnymi operáciami. Na to, aby nejaká štruktúra bola pole, musí spĺňať viacero podmienok (viď Definícia 10) - viac ako grupa. Tým pádom overenie, či je daná štruktúra pole, je o niečo zdĺhavejšie (potreba overiť splnenie viacerých podmienok ako pri grupe). 

Napr. ak $m$ je prvočíslo, tak $\mathbb{Z}_m$ spolu s operáciami $\oplus$ a $\odot$, čo zapisujeme $\left(\mathbb{Z}_m,\oplus,\odot\right)$, má podobné vlastnosti ako množina reálnych čísel $\mathbb{R}$ (resp. racionálnych $\mathbb{Q}$) s obvyklými operáciami $+$ a $\cdot$. Hovoríme, že trojica $\left(\mathbb{Z}_m,\oplus,\odot\right)$ ale aj trojica $\left(\mathbb{R},+,\cdot\right)$ tvoria (matematickú) štruktúru nazývanú _pole_. 

Ak by sme miesto množín $\mathbb{Z}_m$ resp. $\mathbb{R}$ zobrali vo všeobecnosti nejakú množinu $X$ (množina funkcií, čísel, zvyškových tried, matíc, atď.) a miesto operácii $\oplus$, $\odot$ resp. $+$,$\cdot$ uvažovali nejaké binárne operácie definované na množine $X$ (operácie medzi dvoma prvkami množiny $X$, napr. sčítanie a násobenie funkcií, čísel, zvyškových tried alebo matíc), tak môžeme vysloviť všeobecnú definíciu _poľa_, ktorej $\left(\mathbb{Z}_m,\oplus,\odot\right)$ a $\left(\mathbb{R},+,\cdot\right)$ sú špeciálne prípady. 

---

### <font color=red> Definícia 10 </font>

Nech $X$ je ľubovoľná množina a nech $\boxplus$, $\boxdot$ sú binárne operácie na $X$, ktoré sú uzavreté vzhľadom na $X$ (t.j. $X\times X\rightarrow X$). Trojica $\left(X,\boxplus,\boxdot\right)$ sa nazýva **_pole_**, ak platí: 

1. komutatívnosť pre $\boxplus$: $\left(\forall a,b\in X\right)\,a\boxplus b=b\boxplus a$;


2. asociatívnosť pre $\boxplus$: $\left(\forall a,b,c\in X\right)\,\left(a\boxplus b\right)\boxplus c=a\boxplus\left(b\boxplus c\right)$;


3. existencia nulového prvku pre $\boxplus$: $\left(\exists \boldsymbol{0}\in X\right)\left(\forall a\in X\right)\,a\boxplus\boldsymbol{0}=a$;


4. existencia opačného ("záporného") prvku pre $\boxplus$: $\left(\forall a\in X\right)\left(\exists b\in X\right)\,a\boxplus b=\boldsymbol{0}$;


5. komutatívnosť pre $\boxdot$: $\left(\forall a,b\in X\right)\,a\boxdot b = b\boxdot a$;


6. asociatívnosť pre $\boxdot$: $\left(\forall a,b,c\in X\right)\,\left(a\boxdot b\right)\boxdot c = a\boxdot\left(b\boxdot c\right)$;


7. existencia jednotkového prvku pre $\boxdot$: $\left(\exists \boldsymbol{1}\in X\right)\left(\forall a\in X\right)\,a\boxdot\boldsymbol{1}=a$;


8. existencia inverzného prvku ("prevrátenej hodnoty") pre $\boxdot$ $\left(\forall a\in X,a\neq\boldsymbol{0}\right)\left(\exists b\in X\right)\,a\boxdot b=\boldsymbol{1}$;


9. distributívnosť $\boxdot$ vzhľadom na $\boxplus$ zľava: $\left(\forall a,b,c\in X\right)\,a\boxdot\left(b\boxplus c\right)=\left(a\boxdot b\right)\boxplus\left(a\boxdot c\right)$;


10. distributívnosť $\boxdot$ vzhľadom na $\boxplus$ sprava: $\left(\forall a,b,c\in X\right)\,\left(a\boxplus b\right)\boxdot c=\left(a\boxdot c\right)\boxplus\left(b\boxdot c\right)$;

---

---

#### Poznámka 9

Uzavretosťou binárnej operácie (viď Poznámka 1), napr. operácie $\oplus$, sa myslí to, že keď vezmeme dva prvky $a,b\in X$, aplikujeme na nich danú binárnu operáciu $a\oplus b = c$, potom výsledok tejto operácie, t.j. prvok $c$ bude tiež patriť do množiny $X$ tzn. $c\in X$. Zvykne sa to zapisovať štýlom $\oplus:X\times X\rightarrow X$. Inými slovami, binárna operácia je uzavretá vzhľadom na nejakú množinu $X$, keď výsledok tejto binárnej operácie medzi dvoma prvkami z množiny $X$ tiež patrí do množiny $X$. 

---

---

#### Poznámka 10

Pole, podobne ako grupa je z hľadiska abstraktnej algebry veľmi dôležitá štruktúra. Avšak napr. aj pri zavádzaní matíc alebo polynómov resp. pri práci s nimi, ste sa mohli stretnúť s formulkami typu "nech $\mathbf{M}$ je matica nad poľom reálnych (prípadne komplexných) čísel" alebo "uvažujme polynómy štvrtého stupňa nad poľom racionálnych čísel" a pod. Pri narábaní s rôznymi matematickými objektmi je dôležité vedieť, nad akým poľom pracujeme lebo od toho potom závisia určité vlastnosti týchto objekov a vzťahy medzi nimi.
