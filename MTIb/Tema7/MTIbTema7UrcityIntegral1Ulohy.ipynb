**5. Rýchlosť prítoku vody do nádrže je $r(t)$ galónov za minútu, kde $t$ je čas v minútach.** 

**a) Napíšte výraz aproximujúci množstvo vody pritečenej do nádrže v čase od $t$ do $t + \Delta t$, kde $\Delta t$ je malé.** 

**b) Napíšte Riemannov súčet aproximujúci celkové množstvo vody, ktoré pritečie do nádrže v intervale od $t=0$ po $t=5$. Napíšte tiež výraz, ktorý bude presne vyjadrovať toto množstvo.** 

**c) Ako sa zmení množstvo vody v nádrži v intervale od $t=0$ po $t=5$ ak $r(t)=20e^{0.02t}$ ?** 

**d) Ak je $r(t)$ dané vzťahom ako v úlohe c) a ak predpokladáme, že na začiatku je v nádrži 3000 galónov vody, nájdite vzorec $Q(t)$, ktorý bude vyjadrovať množstvo vody v nádrži v čase $t$.**
