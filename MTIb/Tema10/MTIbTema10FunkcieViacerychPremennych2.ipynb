Začnime s pojmom limity. V princípe ide o ten istý koncept ako pri funkciách jednej reálnej premennej ale s tým rozdielom, že teraz máme vo všeobecnosti nekonečne veľa možností, ako sa môžeme blížiť k nejakému bodu $[x_0,y_0]$ v rovine. Uvažujme pre jednoduchosť počiatok súradnicovej sústavy t.j. bod $[x_0,y_0]=[0,0]$. Zatiaľ čo pri funkciách jednej reálnej premennej sme sa k bodu $x_0=0$ mohli blížiť len po osi $x$ a to zľava alebo sprava, v rovine je situácia o čosi zložitejšia (viď obrázok nižšie). V rovine sa môžeme k bodu $[x_0,y_0]=[0,0]$ blížiť napr. po osi $x$ zľava či sprava, alebo po osi $y$ zhora či zdola, alebo po priamke $y=x$ či po priamke $y=-x$ alebo po akejkoľvek inej krivke v rovine, ktorá prechádza bodom $[0,0]$, napr. to môže byť krivka $y=x^2$ alebo $y=-x^2$ a pod.

Význam a zmysel limity je však rovnaký ako ho poznáme pri funkciách jednej reálnej premennej, t.j. limita aj tu vyjadruje správanie sa funkcie $f(x,y)$ v okolí (v blízkosti) nejakého bodu $[x_0,y_0]$, prípadne môže niektorá z premenných smerovať do nekonečna. Inými slovami, zjednodušene povedané, hovoríme, že funkcia $f(x,y)$ má v bode $[x_0,y_0]$ limitu rovnú hodnote $L$ ak pre všetky body $[x,y]$ z nejakého (dostatočne malého prstencového) okolia bodu $[x_0,y_0]$ platí, že aj funkčné hodnoty $f(x,y)$ sú blízko hodnoty $L$ a zapisujeme to v tvare 

$$\displaystyle\lim\limits_{[x,y]\to[x_0,y_0]}f(x,y)=L.$$ 

Tieto úvahy sú ilustrované na nasledujúcom obrázku a Definícia 1 formalizuje uvedené myšlienky ohľadom limity funkcie dvoch reálnych premenných. 


$$\text{Zdroj: Hartman et al. (2018).}$$


---

### <font color=red> Definícia 1 </font> 

Hovoríme, že funkcia $f(x,y)$ má v bode $[x_0,y_0]$ limitu rovnú hodnote $L$, ak 

$$\Big(\forall\varepsilon>0\Big)\Big(\exists\delta>0\Big)\Big(\forall[x,y]:\,0<\big|\big|[x,y]-[x_0,y_0]\big|\big|<\delta\Big)\,\big|f(x,y)-L\big|<\varepsilon.$$

---

#### Poznámka 1 

* Všimnime si, že tak ako v prípade funkcie jednej reálnej premennej ani teraz limita nezávisí od funkčnej hodnoty v skúmanom bode $[x_0,y_0]$. 


* Pri funkciách jednej premennej sme za okolie bodu $x_0$ považovali otvorený interval obsahujúci bod $x_0$ a za prstencové okolie otvorený interval okolo bodu $x_0$ ale bez samotného bodu $x_0$. Pri funkciách dvoch premenných rozumieme okolím bodu $[x_0,y_0]$ kruh (bez kružnice) obsahujúci bod $[x_0,y_0]$ a prstencovým okolím bodu $[x_0,y_0]$ rozumieme kruh (bez kružice) okolo bodu $[x_0,y_0]$ ale bez samotného bodu $[x_0,y_0]$. Ak má kruh polomer $\delta$, tak hovoríme o $\delta$-okolí bodu $[x_0,y_0]$. 


* Definícia 1 teda hovorí, že $L$ je limitou funkcie $f$ v bode $[x_0,y_0]$, ak existuje taký kruh so stredom v bode $[x_0,y_0]$ a s polomerom $\delta>0$, že pre všetky body $[x,y]$ z tohto kruhu (okrem kružnice a bodu $[x_0,y_0]$) platí, že funkčné hodnoty $f(x,y)$ v týchto bodoch sú od hodnoty $L$ vzdialené o menej ako $\varepsilon$, pričom Definícia 1 hovorí, že ku každému $\varepsilon>0$ musí existovať kruh s nejakým polomerom $\delta>0$, aby uvedená podmienka bola splnená. 


* Zatiaľ čo pri funkciách jednej premennej sme pre výpočet vzdialenosti dvoch bodov $x$ a $x_0$ z reálnej osi $\mathbb{R}$ využívali absolútnu hodnotu $|x-x_0|$, pri funkciách dvoch premenných využívame pre výpočet vzdialenosti dvoch bodov $[x,y]$ a $[x_0,y_0]$  z roviny $\mathbb{R}^2$ vzorec z témy 6 v MTIa t.j. $\big|\big|[x,y]-[x_0,y_0]\big|\big|=\sqrt{(x-x_0)^2+(y-y_0)^2}$. 


* Definíciu 1 je možné prirodzene rozšíriť aj pre funkciu troch reálnych premenných, avšak v tom prípade je potrebné využiť príslušný vzorec pre výpočet vzdialenosti dvoch bodov v 3D a okolia bodov už nebudú kruhy ale vnútra gulí v 3D. Túto definíciu je možné pomerne jednoducho zovšeobecniť aj pre funkcie $n$ reálnych premenných.

#### Poznámka 2 

Ak existuje $\lim\limits_{[x,y] \to [x_0,y_0]} f(x,y)$, tak pre každú rovinnú krivku $K$ takú, že každý jej bod (s možnou výnimkou bodu $[x_0,y_0]$) patrí definičnému oboru $f$ a $[x_0,y_0] \in K$ platí
	
$$\lim\limits_{[x,y] \to [x_0,y_0]} f(x,y) = \lim\limits_{{[x,y] \to [x_0,y_0]} \atop {[x,y] \in K}} f(x,y).$$ 

Ináč povedané, keď to zjednodušíme, ak existuje limita funkcie $f(x,y)$ v bode $[x_0,y_0]$, tak potom táto limita musí byť rovnaká nech sa blížime do bodu $[x_0,y_0]$ po akejkoľvek krivke v rovine, ktorá obsahuje bod $[x_0,y_0]$. 

Táto vlastnosť limity nám dáva návod k tomu, ako je možné veľakrát ukázať, že limita nejakej funkcie $f(x,y)$ v bode $[x_0,y_0]$ neexistuje. Stačí totiž nájsť dve vhodné krivky (priamku, parabolu, ...), vypočítať limity pozdĺž týchto kriviek a ak dostaneme dve rôzne limity, potom $\lim\limits_{[x,y] \to [x_0,y_0]} f(x,y)$ neexistuje.

#### <font color=blue> Príklad 1 </font> 

Dokážte, že limita $\lim\limits_{[x,y] \to [0,0]} \dfrac{x^3-y}{x^3+y}$ neexistuje. 

<ins> Riešenie: </ins>

Nech $K_1$ je priamka $y=0$ a $K_2$ je priamka $x=0$. Potom platí
	
$$\lim\limits_{{[x,y] \to [0,0]} \atop {[x,y] \in K_1}} \dfrac{x^3-y}{x^3+y} = \lim\limits_{{[x,y] \to [0,0]} \atop {y=0}} \dfrac{x^3-y}{x^3+y} = \lim\limits_{x \to 0} \dfrac{x^3-0}{x^3+0} = 1$$

$$\lim\limits_{{[x,y] \to [0,0]} \atop {[x,y] \in K_2}} \dfrac{x^3-y}{x^3+y} = \lim\limits_{{[x,y] \to [0,0]} \atop {x=0}} \dfrac{x^3-y}{x^3+y} = \lim\limits_{y \to 0} \dfrac{0-y}{0+y} = -1$$

Keďže pre dve rôzne krivky $K_1,K_2$ sa výsledky limít nerovnajú, limita neexistuje.
 

```python
# kontrola limit pomocou Sage 
x,y = var('x,y')
f(x,y) = (x^3-y) / (x^3+y)
show(f)
show(limit(f(x,0),x=0)) # limita pre x->0 pricom y=0
show(limit(f(0,y),y=0)) # limita pre y->0 pricom x=0
```

Samozrejme, zadaná funkcia nie je definovaná v bode $[0,0]$, keďže po dosadení vznikne nula v menovateli zlomku. Napriek tomu by mohla mať v tomto bode limitu, avšak ako ukazuje predošlý výpočet a nasledujúci obrázok, limita zadanej funkcie v bode $[0,0]$ neexistuje. Na obrázku nižšie sú znázornené červenou farbou funkčné hodnoty pozdĺž priamky $y=0$, ktoré sú vo výške $z=1$ v okolí bodu $[0,0]$, naproti tomu funkčné hodnoty (modrou farbou) pozdĺž priamky $x=0$ sú vo výške $z=-1$ v okolí bodu $[0,0]$, zjednodušene povedané, červená krivka je nad modrou krivkou v okolí bodu $[0,0]$ t.j. krivky sa rozchádhajú a teda limita zadanej funkcie v bode $[0,0]$ nemôže existovať. 

```python
x,y = var('x,y')
f(x,y) = (x^3-y) / (x^3+y) 
p = plot3d(f,(x,-5,5),(y,-5,5), color='grey', mesh=True) # vykreslenie grafu (plochy) 
p += parametric_plot3d((x, 0, f(x,0)), (x,-5,5), color='red') # pridame krivku (cast priamky y=0) do grafu 
p += parametric_plot3d((0, y, f(0,y)), (y,-5,5), color='blue') # pridame krivku (cast priamky x=0) do grafu
show(p) # vykreslime plochu spolu s krivkami  
```

### <font color=green> Veta 1 </font> 

Nech $b,x_0,y_0,L,K$ sú reálne čísla a nech $n$ je číslo prirodzené. Majme dve funkcie $f,g$ s týmito limitami: 

$$\lim\limits_{[x,y] \to [x_0,y_0]} f(x,y)=L \qquad \text{ a } \qquad \lim\limits_{[x,y] \to [x_0,y_0]} g(x,y)=K.$$ 

Potom platia nasledovné vzťahy: 

1. $\lim\limits_{[x,y] \to [x_0,y_0]} b = b$; 


2. $\lim\limits_{[x,y] \to [x_0,y_0]} x = x_0$ a $\lim\limits_{[x,y] \to [x_0,y_0]} y = y_0$; 


3. $\lim\limits_{[x,y] \to [x_0,y_0]} \left(f(x,y)\pm g(x,y) \right)=L\pm K$; 


4. $\lim\limits_{[x,y] \to [x_0,y_0]} b\cdot f(x,y)=bL$; 


5. $\lim\limits_{[x,y] \to [x_0,y_0]} f(x,y)\cdot g(x,y)=LK$; 


6. ak $K\neq0$, tak $\lim\limits_{[x,y] \to [x_0,y_0]} \dfrac{f(x,y)}{g(x,y)}=\dfrac{L}{K}$; 


7. $\lim\limits_{[x,y] \to [x_0,y_0]} \left(f(x,y)\right)^n=L^n$.

* Platí napr. aj veta "$0$ krát ohraničená funkcia", podobne ako tomu bolo aj pri funkcii jednej premennej.

<ins> Riešenie: </ins>

#### <font color=blue> Príklad 4 </font>  

Vypočítajte $\lim\limits_{[x,y] \to [0,0]} \dfrac{x^2y+xy^2}{x^2-xy+y^2}$.  

<ins> Riešenie: </ins> 

Funkciu $f(x,y)=\dfrac{x^2y+xy^2}{x^2-xy+y^2}$ vyjadríme v tvare súčinu dvoch funkcií, z ktorých jedna má limitu $0$ a druhá je ohraničená: 

$$f(x,y)=\dfrac{x^2y+xy^2}{x^2-xy+y^2}=\dfrac{xy(x+y)}{x^2-xy+y^2}=(x+y)\dfrac{xy}{x^2-xy+y^2}.$$

Pretože funkcia $g(x,y)=x+y$ je spojitá resp. je to polynomická funkcia (vzhladom na premenné $x,y$), jej limita sa rovná jej hodnote v bode $[0,0]$, čiže 

$$\lim\limits_{[x,y] \to [0,0]}g(x,y)=g(0,0)=0+0=0.$$

Ak sa nám o funkcii $h(x,y)=\dfrac{xy}{x^2-xy+y^2}$ podarí ukázať, že je ohraničená, tak môžeme aplikovať vetu "nula krát ohraničená" pre výpočet zadanej limity. 

Keďže platí 

$$2\left(x^2-xy-y^2\right)=2x^2-2xy+2y^2=x^2+x^2-2xy+y^2+y^2=x^2+(x-y)^2+y^2,$$

odkiaľ máme 

$$x^2-xy-y^2=\frac{x^2+(x-y)^2+y^2}{2},$$

preto výraz v menovateli funkcie h nemôže byť záporný. Tiež odtiaľ vyplýva, že funkcia $h$ je definovaná pre každé $[x,y]\neq[0,0]$. Všimnime si, že pre každé $x\neq0$ platí: 

$$h(x,x)=1, \qquad h(x,-x)=-\frac{1}{3}.$$

Ukážeme, že pre každé $[x,y]\neq[0,0]$ je funkcia $h$ ohraničená, konkrétne teda ukážeme, že platí 

$$-\frac{1}{3}\le\dfrac{xy}{x^2-xy+y^2}\le1.$$

Pretože $x^2-xy+y^2>0$, nasledujúce úpravy sú ekvivalentné 

$$\dfrac{xy}{x^2-xy+y^2}\le1$$

$$xy\le x^2-xy+y^2$$ 

$$0\le x^2-2xy+y^2$$

$$0\le(x-y)^2.$$

Z rovnakého dôvodu sú aj nasledujúce úpravy ekvivalentné: 

$$-\frac{1}{3}\le\dfrac{xy}{x^2-xy+y^2}$$

$$-x^2+xy-y^2\le 3xy$$

$$0\le x^2+2xy+y^2$$

$$0\le(x+y)^2.$$

Tým je ohraničenosť funkcie $h$ dokázaná a podľa spomínanej vety dostávame 

$$\lim\limits_{[x,y] \to [0,0]} \dfrac{x^2y+xy^2}{x^2-xy+y^2}=0.$$

Ďalej ešte uvádzame graf funkcie $h$, z ktorého by sme tiež mohli odhadnúť horné i dolné ohraničenie (podľa max a min hodnoty na osi $z$) a následne sa pokúsiť vhodnými úpravami (ako sme to urobili vyššie) dokázať, že zvolené ohraničenia sú v poriadku. 

```python
x, y = var('x,y')  
h(x,y) = (x*y) / (x^2-x*y+y^2) 
plot3d(h,(x,-10,10),(y,-10,10), color='grey', opacity=0.8)
```

Aj koncept **spojitosti funkcie** má svoju analógiu pri funkciách dvoch a viacerých premenných. Interpretácia je veľmi podobná ako v prípade funkcie jednej reálnej premennej. Uvažujme nasledovný príklad pre ilustráciu spojitosti funkcie dvoch premenných. 

Strmá (kolmá) stena resp. údolie Half Dome v kalifornskom národnom parku Yosemite bola spôsobená ľadovou aktivitou počas doby ľadovej (pozri obrázok nižšie). Ak sa približujeme po skale od západu, výška terénu náhle stúpne o takmer 5 000 stôp od "podlahy" údolia, 2 000 stôp od neho vertikálne. <br/>
Ak vezmeme do úvahy funkciu $h$ udávajúcu výšku terénu nad morom v závislosti od zemepisnej dĺžky a šírky, potom má $h$ nespojitosť (skoky) pozdĺž cesty na úpätí útesu Half Dome. Pri pohľade na vrstevnicovú mapu oblasti na obrázku nižšie (vpravo) vidíme, že na väčšine miest má malá zmena polohy za následok malú zmenu výšky (čo odpovedá spojitosti funkcie $h$), s výnimkou blízkosti útesu. Tam, nech urobíme akýkoľvek malý krok, nastane veľká zmenu výšky (to odpovedá nespojitosti funkcie $h$ a  môžete vidieť, ako sa zhustené vrstevnice dostávajú k útesu; niektoré sa náhle končia pozdĺž nespojitosti). <br/>
Tento geologický útvar ilustruje myšlienky spojitosti a nespojitosti. Zhruba povedané, o funkcii hovoríme, že je spojitá v danom bode, ak sa jej funkčné hodnoty na miestach v blízkosti daného bodu blížia k funkčnej hodnote v danom bode. Ak to tak nie je, hovoríme o funkcii nespojitej. 

 
$$\text{Zdroj: Hallet et al. (2012).}$$

Vlastnosť spojitosti je taká vlastnosť funkcií, ktorú zvyčajne predpokladáme o funkciách, ktoré študujeme. Zjednodušene povedané, očakávame (s výnimkou zvláštnych okolností), že hodnoty funkcie sa pri malých zmenách vstupných premenných drasticky nezmenia. Kedykoľvek modelujeme funkciu jednej premennej pomocou neprerušenej krivky, robíme tento predpoklad. Aj keď nám funkcie prídu ako tabuľky údajov, zvyčajne vychádzame z predpokladu, že chýbajúce hodnoty funkcií medzi údajmi sú blízke tým  nameraným. <br/>
V tejto časti študujeme limity a spojitosť trochu formálnejšie v kontexte funkcií viacerých premenných. Pre jednoduchosť študujeme tieto koncepty pre funkcie dvoch premenných, ale našu diskusiu je možné prispôsobiť funkciám troch premenných alebo zovšeobecniť pre funkcie viacerých premenných. <br/>
Je možné ukázať (Veta 2), že súčty, súčiny a zloženie spojitých funkcií sú spojité, zatiaľ čo podiel dvoch spojitých funkcií je spojitý všade tam, kde je funkcia menovateľa nenulová. Teda napr. každá z funkcií

$$\cos\left(x^2y\right), \qquad \ln\left(x^2+y^2\right), \qquad \frac{e^{x+y}}{x+y}, \qquad \ln\left(\sin\left(x^2+y^2\right)\right)$$

je spojitá vo všetkých bodoch $[x,y]$ roviny $\mathbb{R}^2$, v ktorých je definovaná. Podobne ako pri funkcii jednej premennej, tak aj tu graf spojitej funkcie nad neprerušenou doménou resp. časťou roviny (ako napr. kruh, štvorec, polrovina a pod.) je neprerušený - to znamená, že povrch resp. plocha (graf funkcie v 3D) nemá v sebe žiadne "diery" ani "trhliny".

### <font color=red> Definícia 2 </font> 

Nech funkcia $f(x,y)$ je definovaná na nejakej množine $D\subset\mathbb{R}^2$ (okolí) obsahujúcej bod $[x_0,y_0]$. 

1. Funkcia $f$ je **_spojitá v bode_** $[x_0,y_0]$ ak platí $\lim\limits_{[x,y]\to[x_0,y_0]}f(x,y)=f(x_0,y_0)$. 


2. Funkcia $f$ je **_spojitá na množine_** $D$ ak je spojitá v každom bode množiny $D$. Ak je funkcia $f$ spojitá vo všetkých bodoch roviny $\mathbb{R}^2$, potom hovoríme, že $f$ je **_spojitá všade_**.

# <font color=brown> Parciálne derivácie </font>

Nech $y$ je funkcia $x$. Podrobne sme študovali deriváciu $y$ vzhľadom na $x$, t.j. $\frac{\mathrm{d}y}{\mathrm{d}x}$, ktorá určuje rýchlosť (intenzitu), akou sa $y$ mení vzhľadom na $x$. Uvažujme teraz $z = f (x, y)$. V tomto prípade má zmysel skúmať, ako sa $z$ mení vzhľadom na $x$ a/alebo $y$. Tieto miery/intenzity/rýchlosti zmien funkcie $f$ budeme skúmať v poslednej časti učebného textu. 

Uvažujme funkciu $z = f (x, y) = x^2 + 2y^2$, ako je znázornené na obrázku nižšie. Fixovaním $y = 2$ zameriavame našu pozornosť  na všetky body na povrchu, ktoré majú súradnicu $y$ rovnú hodnote $2$ (viď obrázok). Tieto body tvoria krivku (modrou farbou) v priestore s rovnicou $z = f (x, 2) = x^2 + 8$, ktorá je funkciou iba jednej premennej (rez funkcie). Pozdĺž tejto krivky môžeme uvažovať derivácie $z$ vzľadom na $x$ a nájsť napr. rovnice dotyčníc atď. V tomto prípade je parciálna derivácia funkcie $f$ podľa premennej $x$ vlastne smernica dotyčnice k funkcii (krivke) $z = f (x, 2) = x^2 + 8$ vo zvolenom bode. 

 
$$\text{Zdroj: Hartman et al. (2018).}$$

Kľúčová predstava, ktorú je potrebné si vziať z tohto príkladu je, že považovaním $y$ za konštantné (nemení sa) môžeme uvažovať o tom, ako sa funkcia $z$ mení vzhľadom na $x$. Podobným spôsobom môžeme ponechať konštantné $x$ a skúmať, ako sa funkcia $z$ mení vzhľadom na $y$. Toto je základný princíp _parciálnych derivácií_. Najprv uvedieme formálnu definíciu založenú na limitách a potom ukážeme, ako vypočítať tieto parciálne derivácie aj bez priameho použitia limít. 

---

### <font color=red> Definícia 3 </font> 

Nech funkcia $z=f(x,y)$ je definovaná na nejakom okolí bodu $A=[x_0,y_0]$. Hovoríme, že funkcia $f$ má **_v bode $A$ parciálnu deriváciu podľa premennej $x$_**, ak existuje limita

$$\left.\frac{\partial f}{\partial x} \right |_{A} = \lim\limits_{x \to x_0} \dfrac{f(x,y_0) - f(x_0,y_0)}{x - x_0} = \lim\limits_{h \to 0} \dfrac{f(x_0+h,y_0) - f(x_0,y_0)}{h}.$$

Taktiež hovoríme, že funkcia $f$ má v bode $A$ **_parciálnu deriváciu podľa premennej $y$_**, ak existuje limita

$$\left.\frac{\partial f}{\partial y} \right |_{A} = \lim\limits_{y \to y_0} \dfrac{f(x_0,y) - f(x_0,y_0)}{y - y_0} = \lim\limits_{h \to 0} \dfrac{f(x_0,y_0+h) - f(x_0,y_0)}{h}.$$

Analogické označenia sa používajú aj pre parciálnu deriváciu podľa premennej $y$ v bode $A$.

### <font color=red> Definícia 4 </font> 

**Parciálna derivácia funkcie $z=f(x,y)$ podľa premennej $x$ je funkcia** $F(x,y)$, ktorej definičný obor je množina $\cal{D}\subset\mathbb{R}^2$ všetkých bodov, v ktorých má $f$ parciálnu deriváciu podľa $x$ a ktorej hodnota sa v každom bode $A \in \cal{D}$ rovná parciálnej derivácii v $A$ podľa $x$; označujeme ju $\dfrac{\partial f}{\partial x}$, $f_x$ alebo $\dfrac{\partial z}{\partial x}$ či $z_x$ a pod. Analogicky definujeme a označíme aj **parciálnu deriváciu funkcie $z=f(x,y)$ podľa premennej $y$**.

Pre funkciu $f(x,y)=x^2y+2x+y^3$ určte jej parciálnu deriváciu $f_x(x,y)$ z definície (t.j. pomocou limity).

Parciálne derivácie je možné analogicky definovať aj pre funkciu troch či viacerých - vo všebecnosti teda $n$ reálnych premenných. Ich význam ostáva rovnaký, problematická je však ich vizualizácia. Definícia parciálnej derivácie danej funkcie $f(x_1,\ldots,x_n)$ podľa premennej $x_i$ v konkrétnom bode $A=[a_1,\ldots,a_n]$ je nasledovná: 

---

### <font color=red> Definícia 5 </font> 

Nech $f$ je funkcia $n \geq 2$ premenných $x_1,\dots,x_n$ definovaná na nejakom okolí bodu $A = [a_1,\dots, a_n]$. Hovoríme, že
funkcia $f$ má **_v bode $A$ parciálnu deriváciu podľa premennej $x_i$_**, ak existuje limita

$$
\lim\limits_{x_i \to a_i} \frac{f(a_1,\dots, a_{i-1},x_i,a_{i+1},\dots,a_n) - f(a_1,\dots, a_{i-1},a_i,a_{i+1},\dots,a_n)}{x_i - a_i}=\lim\limits_{h \to 0} \frac{f(a_1,\dots, a_{i-1},a_i+h,a_{i+1},\dots,a_n) - f(a_1,\dots, a_{i-1},a_i,a_{i+1},\dots,a_n)}{h}.$$ 

Hodnotu tejto limity označujeme $\left.\dfrac{\partial f}{\partial x_i} \right |_A$ alebo $f_{x_i}(A)$. 

--- 

---

### <font color=red> Definícia 6 </font> 

**_Parciálna derivácia funkcie $f$ podľa premennej $x_i$_** je funkcia $F(x_1,\dots, x_n)$, ktorej definičný obor je množina $\cal{D}$ všetkých bodov, v ktorých má $f$ parciálnu deriváciu podľa $x_i$ a ktorej hodnota sa v každom bode $A \in \cal{D}$ rovná parciálnej derivácii v $A$ podľa $x_i$; označujeme ju $\frac{\partial f}{\partial x_i}$ alebo $f_{x_i}$.

Keďže parciálne derivácie $f_x(x,y)$ a $f_y(x,y)$ funkcie $f$ sú sami o sebe funkciami premenných $x,y$, tým pádom prirodzene tak, ako to bolo aj pri funkciách jednej premennej, aj tu má zmysel robiť derivácie vyšších rádov. Naskytá sa tu ale viacero možností ako urobiť parciálne derivácie napr. druhého rádu. Môžeme najprv derivovať zadanú funkciu $f$ podľa $x$, čím vznikne $f_x$ a následne môžeme derivovať podľa $y$ čím vznikne $f_{xy}$ alebo môžeme derivovať $f_x$ podľa $x$, čím vznikne $f_{xx}$ atď. Ukážeme si to na konkrétnom príklad, ktorý nasleduje za definíciou parciálnych derivácií druhého rádu.

### <font color=red> Definícia 7 </font> 

Nech $z=f(x,y)$ je funkcia definovaná na nejakej množine $\cal{D}\subset\mathbb{R}^2$. **_Parciálnu deriváciu druhého rádu podľa $x$ a následne podľa $x$_** definujeme vzťahom 

$$f_{xx}=\dfrac{\partial}{\partial x}\left(\dfrac{\partial f}{\partial x}\right)=\dfrac{\partial^2f}{\partial x^2}.$$

**_Parciálnu deriváciu druhého rádu podľa $x$ a následne podľa $y$_** definujeme vzťahom 

$$f_{xy}=\dfrac{\partial}{\partial y}\left(\dfrac{\partial f}{\partial x}\right)=\dfrac{\partial^2f}{\partial y \partial x}.$$

Analogicky definujeme $f_{yy}=\dfrac{\partial^2f}{\partial y^2}$ a $f_{yx}\dfrac{\partial^2f}{\partial x \partial y}$. 

Parciálne derivácie druhého rádu $f_{xy}$ a $f_{yx}$ sa nazývajú **_zmiešané parciálne derivácie_**.

#### <font color=blue> Príklad 11 </font> 

Je daná funkcia $f(x,y) = \dfrac{xy}{x+y+2}$. Vypočítajte $\dfrac{\partial^2 f}{\partial x^2},\dfrac{\partial^2 f}{\partial y^2},$ $\dfrac{\partial^2 f}{\partial x \partial y}, \dfrac{\partial^2 f}{\partial y \partial x}$. 

<ins> Riešenie: </ins> 

Platí

$$\dfrac{\partial f}{\partial x} = \dfrac{(xy)'(x+y+2) - xy(x+y+2)'}{(x+y+2)^2} = \dfrac{y(x+y+2) - xy\cdot 1}{(x+y+2)^2}=\dfrac{y(y+2)}{(x+y+2)^2};$$
	
$$\dfrac{\partial f}{\partial y} = \dfrac{(xy)'(x+y+2) - xy(x+y+2)'}{(x+y+2)^2} = \dfrac{x(x+y+2) - xy\cdot 1}{(x+y+2)^2}=\dfrac{x(x+2)}{(x+y+2)^2};$$ 
    
$$\dfrac{\partial^2 f}{\partial x^2} = \dfrac{\partial}{\partial x}\left(\dfrac{y(y+2)}{(x+y+2)^2}\right ) = \frac{(y(y+2))'(x+y+2)^2-y(y+2)((x+y+2)^2)'}{(x+y+2)^4}=\dfrac{0\cdot(x+y+2)^2-y(y+2)\cdot 2(x+y+2)\cdot 1}{(x+y+2)^4} = \dfrac{-2y(y+2)}{(x+y+2)^3};$$

$$\dfrac{\partial^2 f}{\partial y^2} = \dfrac{\partial}{\partial y}\left(\dfrac{x(x+2)}{(x+y+2)^2}\right ) = \frac{(x(x+2))'(x+y+2)^2-x(x+2)((x+y+2)^2)'}{(x+y+2)^4}=\dfrac{0\cdot(x+y+2)^2-x(x+2)\cdot 2(x+y+2)\cdot 1}{(x+y+2)^4} = \dfrac{-2x(x+2)}{(x+y+2)^3};$$ 
    
$$\dfrac{\partial^2 f}{\partial x \partial y} = \dfrac{\partial}{\partial x}\left(\dfrac{x(x+2)}{(x+y+2)^2}\right ) =\frac{(2x+2)\cdot(x+y+2)^2-x(x+2)\cdot 2(x+y+2)\cdot 1}{(x+y+2)^4} = \dfrac{2(xy+x+y+2)}{(x+y+2)^3};$$ 

$$\dfrac{\partial^2 f}{\partial y \partial x} = \dfrac{\partial}{\partial x}\left(\dfrac{y(y+2)}{(x+y+2)^2}\right ) =\frac{(2y+2)\cdot(x+y+2)^2-y(y+2)\cdot 2(x+y+2)\cdot 1}{(x+y+2)^4} = \dfrac{2(xy+x+y+2)}{(x+y+2)^3}.$$

\dfrac{\mbox{d}\ \! f}{\mbox{d}\ \!t} = \dfrac{1}{y}\cdot (6t+1)e^{3t^2+t+1} -\dfrac{x}{y^2}\cdot (2t) = \frac{(6t+1)e^{3t^2+t+1}}{t^2-1} - \frac{e^{3t^2+t+1}\cdot (2t)}{(t^2-1)^2} = \dfrac{e^{3t^2+t+1}((6t+1)(t^2-1) - 2t)}{(t^2-1)^2} = \dfrac {{e^{3{t}^{2}+t+1}} \left( 6{t}^{3}-8t+{t}^{2}-1 \right) 
}{ \left( {t}^{2}-1 \right) ^{2}}.$$

### <font color=green> Veta 7 </font> 

Nech $u = u(x,y),v = v(x,y)$ sú funkcie dvoch premenných $x,y$, ktoré majú derivácie v bode $A = [x_0,y_0]$ a nech $f = f(u,v)$ je funkcia dvoch premenných, ktorá má parciálne derivácie podľa $u$ a $v$ v bode $B = [u_0,v_0]$, kde $u_0 = u(x_0,y_0),v_0 = v(x_0,y_0)$. Potom zložená funkcia $f = f(u(x,y),v(x,y))$ má parciálne derivácie v bode $A$ rovné
	
$$\left.\dfrac{\partial f}{\partial x}\right|_{A} = \left .\dfrac{\partial f}{\partial u}\right |_B \cdot \left.\dfrac{\partial u}{\partial  x}\right |_{A} + \left .\dfrac{\partial f}{\partial v}\right |_B \cdot \left.\dfrac{\partial v}{\partial x}\right |_{A}$$ 

$$\left.\dfrac{\partial f}{\partial y}\right|_{A} = \left .\dfrac{\partial f}{\partial u}\right |_B \cdot \left.\dfrac{\partial u}{\partial  y}\right |_{A} + \left .\dfrac{\partial f}{\partial v}\right |_B \cdot \left.\dfrac{\partial v}{\partial y}\right |_{A}$$

#### <font color=blue> Príklad 14 </font> 

Nech $f(x,y) = \mbox{arctg }\! \left ( \dfrac{x+y}{1-xy}\right )$. Vypočítajte $\dfrac{\partial f}{\partial x},\dfrac{\partial f}{\partial y}$ pomocou vzorca pre derivovanie zloženej funkcie. 
	 
<ins> Riešenie: </ins> 

Funkciu $f$ možno vyjadriť tiež ako $f(u,v) = \mbox{arctg }\!\left ( \dfrac{u}{v}\right ), u(x,y) = x+y, v(x,y) = 1-xy$. Platí $\dfrac{\partial f}{\partial u} =  \dfrac{1}{1+\left(\frac{u}{v}\right)^2}\cdot\dfrac{1}{v},$ $\dfrac{\partial f}{\partial v} =  \dfrac{1}{1+\left(\frac{u}{v}\right)^2}\cdot\left (-\dfrac{u}{v^2}\right),$ $\dfrac{\partial u}{\partial x} = \dfrac{\partial u}{\partial y} = 1$, $\dfrac{\partial v}{\partial x} = -y, \dfrac{\partial v}{\partial y} = -x$, teda po dosadení do príslušného vzorca pre derivovanie zloženej funkcie dostávame

$$\dfrac{\partial f}{\partial x} = \left [ \dfrac{1}{1+\left(\frac{u}{v}\right)^2}\cdot\dfrac{1}{v}\right ]\cdot 1 + \left [ \dfrac{1}{1+\left(\frac{u}{v}\right)^2}\cdot\left (-\dfrac{u}{v^2}\right)\right ]\cdot(-y) = \left [ \dfrac{1}{1+\left(\frac{u}{v}\right)^2}\right ] \cdot \left ( \dfrac{1}{v} - \dfrac{u}{v^2}\cdot (-y)\right) =  \dfrac{1}{1+\left(\frac{u}{v}\right)^2}\left ( \dfrac{1}{v} + \dfrac{uy}{v^2} \right ) = \dfrac{1}{1+\left (\frac{x+y}{1-xy}\right)^2}\left (\dfrac{1-xy + (x+y)y}{(1-xy)^2} \right ) = \dfrac{1}{\frac{(1-xy)^2+(x+y)^2}{(1-xy)^2}}\cdot\dfrac{1+y^2}{(1-xy)^2} = \dfrac{1+y^2}{1+x^2+y^2+x^2y^2} = \dfrac{1+y^2}{(1+x^2)(1+y^2)} = \dfrac{1}{1+x^2} $$ 
    
$$\dfrac{\partial f}{\partial y} = \left [ \dfrac{1}{1+\left(\frac{u}{v}\right)^2}\cdot\dfrac{1}{v}\right ]\cdot 1 + \left [ \dfrac{1}{1+\left(\frac{u}{v}\right)^2}\cdot\left (-\dfrac{u}{v^2}\right)\right ]\cdot(-x) = \left [ \dfrac{1}{1+\left(\frac{u}{v}\right)^2}\right ] \cdot \left ( \dfrac{1}{v} - \dfrac{u}{v^2}\cdot (-x)\right) =  \dfrac{1}{1+\left(\frac{u}{v}\right)^2}\left ( \dfrac{1}{v} + \dfrac{ux}{v^2} \right ) = \dfrac{1}{1+\left (\frac{x+y}{1-xy}\right)^2}\left (\dfrac{1-xy + (x+y)x}{(1-xy)^2} \right ) = \dfrac{1}{\frac{(1-xy)^2+(x+y)^2}{(1-xy)^2}}\cdot\dfrac{1+y^2}{(1-xy)^2} = \dfrac{1+x^2}{1+x^2+y^2+x^2y^2} = \dfrac{1+x^2}{(1+x^2)(1+y^2)} = \dfrac{1}{1+y^2} $$
