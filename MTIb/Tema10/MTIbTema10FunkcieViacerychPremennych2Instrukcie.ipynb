# <font color=brown> Motivácia</font>


 

Rôzne **algoritmy** a postupy v odvetviach informatiky **si vyžadujú** určitú **optimalizáciu** (zefektívnenie/spresnenie), ktorá je spojená s **diferenciálnym počtom funkcií viacerých premenných**. Napr. môže ísť o **algoritmus spätnej propagácie v procese trénovania neurónových sietí v rámci umelej inteligencie** alebo **iné učiace algoritmy v strojovom učení** alebo sa môže jednať o **popis deformácií či mapovania textúr v oblasti počítačového videnia resp. grafiky** a pod. 

 

V téme 10 si teda ozrejmíme základné pojmy **diferenciálneho počtu funkcie viacerých premenných**. Konkrétne zavedieme **limitu funkcie** viacerých premenných a pomocou nej, podobne ako pri funkcii jednej premennej, budeme definovať **spojitosť** i  **parciálne derivácie** funkcie viacerých premenných. Názorne si vysvetlíme ich význam - aj geometrický. Ukážeme si, ako vypočítať základné **limity funkcie dvoch premenných v danom bode** resp. ako ukázať, že taká **limita neexistuje**. Taktiež sa naučíme **vypočítať parciálne derivácie** (v danom bode) **funkcie dvoch a troch premenných**.

# <font color=brown> Učebné ciele</font>

### Základné 


* Definovať limitu funkcie dvoch a viacerých premenných. 


* Rozumieť (aj vizuálne), čo to znamená, že funkcia dvoch premenných v danom bode má limitu alebo nemá limitu. 


* Vedieť sformulovať základné pravidlá/tvrdenia pre výpočet limít funkcie dvoch premenných a ilustrovať na konkrétnych príkladoch. 


* Definovať spojitosť funkcie dvoch premenných, rozumieť jej geometrickému významu a vedieť posúdiť spojitosť funkcie na základe vizualizácii. 


* Sformulovať základné pravidlá pre tvorenie spojitých funkcií a vedieť ilustrovať na príklade. 


* Definovať parciálne derivácie funkcie dvoch i viacerých premenných a rozumieť ich významu vrátane geometrického. 


* Definovať a ilustrovať na príklade parciálne derivácie druhého rádu, vedieť za akých podmienok sú zmiešané parcialne derivácie totožné. 


* Rozumieť základným reťazovým pravidlám pre derivovanie zložených funkcií.  

### Pokročilé 


* Vypočítať limitu funkcie dvoch premenných v danom bode s využitím základných pravidiel/tvrdení. 


* Zvolením vhodnej krivky ukázať, že limita funkcie dvoch premenných v danom bode neexistuje. 


* Na základe predpisu a tiež pomocou grafov odhadnúť, či daná funkcia je alebo nie je spojitá na danej množine. 


* Pomocou základných pravidiel (pre funkciu jednej premennej) vypočítať parciálne derivácie zadanej funkcie (v konkrétnom bode). 


* Aplikovaním správneho reťazového pravidla určiť parciálne derivácie zložených funkcií.
