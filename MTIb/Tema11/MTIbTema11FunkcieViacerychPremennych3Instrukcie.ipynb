<a id=motivacia></a>
# <font color=brown> Motivácia</font>


 

V tejto poslednej téme venovanej funkciám viacerých premenných nadviažeme na predchádzajúce dve témy a objasníme ešte niektoré dôležité pojmy i vzťahy medzi nimi. Vysvetlíme si, aký je vzťah medzi **diferencovateľnosťou** a **spojitosťou** funkcie dvoch premenných, pričom poukážeme na podobnosti i rozdiely v porovnaní s funkciami jednej reálnej premennej. V tejto súvislosti budeme definovať **totálny diferenciál** funkcie dvoch premenných a budeme ilustrovať jeho využitie pri **aproximácii** funkčných hodnôt. <br/>
Nadväzujúc na to sa budeme zaoberať aj **dotyčnicami**, **dotykovou rovinou** a **gradient vektorom**, pričom ich význam a interpretáciu si ukážeme na príkladoch. <br/>
Posledná časť bude venovaná hľadaniu **lokálnych extrémov** funkcií dvoch reálnych premenných a ich využitiu **pri riešení optimalizačných úloh**.

### Základné

* Sformulovať vzťah diferencovateľnosti a spojitosti funkcie dvoch reálnych premenných (ilustrovať na príklade). 


* Definovať a interpretovať dotykovú rovinu funkcie dvoch premenných. 


* Definovať totálny diferenciál funkcie dvoch premenných, vedieť ho interpretovať a dať do súvisu s dotykovou rovinou k funkcii v danom bode. 


* Definovať a učiť gradient vektor (v konkrétnom bode) funkcie dvoch premenných, rozumieť jeho významu a interpretácii. 


* Definovať kritický bod, lokálne extrémy a sedlový bod funkcie dvoch premenných. Uviesť príklady.

### Pokročilé

* Využiť totálny diferenciál pri aproximácii hodnôt funkcií dvoch premenných. 


* Nájsť gradient vektor (v konkrétnom bode) zadanej funkcie dvoch premenných a určiť jeho normu. 


* Na základe grafov i vrstevnicových diagramov funkcie dvoch premenných vedieť klasifikovať jej kritické body.
