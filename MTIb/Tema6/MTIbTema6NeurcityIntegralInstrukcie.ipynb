<a id=motivacia></a>
# <font color=brown> Motivácia</font>


 

O deriváciách funkcií už čo-to vieme. S tým veľmi úzko súvisí ďalší pojem - (neurčitý) **integrál**. 

**Analýza algoritmov** sa týka skúmania výpočtovej zložitosti algoritmu. Môže zahŕňať priestor a čas (okrem iných zdrojov), potrebných na vykonanie daného algoritmu. V procese analýzy algoritmov ide zvyčajne o nájdenie funkcií, ktoré porovnávajú čas trvania algoritmu s množstvom potebné úložného priestoru (časová vs pamäťová zložitosť algoritmov). Niekedy sa to robí použitím **integrálneho počtu**. Konkrétne môže ísť o analýzu kombinatorických algoritmov, pravdepodobnostných algoritmov resp. algoritmov využívajúcich prvky náhodnosti ako napr. generovanie pseudonáhodných čísel a pod. <br/>
Ďalej napr. **úloha určenia** (spočítania) **množiny všetkých riešení daného problému** sa rieši s využitím tzv. vytvárajúcej  funkcie, ktorej asymptotické (limitné) správanie sa skúma aj pomocou **integrálnych techník** (napr. u tzv. "zhutnených" (_compacted_) binárnych stromov). 

 

V téme 6 sa teda budeme venovať **procesu**, ktorý je v istom zmysle **opačný k procesu derivovania**, pôjde o **integrovanie**. Na reálnych príkladoch budeme ilustrovať zmysel **integrovania** a uvidíme, že táto operácia rieši úlohy, ktoré vznikajú úplne prirodzene. Zavedieme pojem **primitívnej funkcie** (**antiderivácie**) a následne definujeme **neurčitý integrál** spolu s jeho vlastnosťami. Poukážeme aj na to, že nie ku každej funkcii je možné nájsť **primitívnu funkciu**, ktorá by sa dala vyjadriť pomocou elementárnych funkcií. <br/>
V ďalšej časti učebného textu si ukážeme na príkladoch, ako vypočítať **neurčité integrály** z funkcií za pomoci tzv. **vzorcov pre integrovanie**. Vysvetlíme a na príkladoch budeme ilustrovať ako fungujú dve štandardné metódy pre integrovanie - **substitučná metóda** a **metóda per partes**.
